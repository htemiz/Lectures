# Metin Alanı (Text Field) Oluşturma ve Şekillendirme

Metin alanları, kullanıcıların uygulamaya yazı yazmasına olanak tanır.
Formlar oluşturmak, mesaj göndermek, arama deneyimleri yaratmak ve daha
fazlası için kullanılırlar.

Flutter bu işlem için iki temel widget sunar: **`TextField`** ve
**`TextFormField`**.

------------------------------------------------------------------------

## 1. TextField

`TextField`, en yaygın kullanılan metin giriş widget’ıdır.

- **Varsayılan Görünüm:** Varsayılan olarak `TextField`’ın altı
  çizilidir.
- **Özelleştirme:** `decoration` özelliğine bir **`InputDecoration`**
  nesnesi vererek etiket (label), simge (icon), satır içi ipucu metni
  (hint text) ve hata metni ekleyebilirsiniz.
- **Sadeleştirme:** Süslemeyi (alt çizgi ve etiket için ayrılan alan
  dahil) tamamen kaldırmak için `decoration` özelliğini `null` olarak
  ayarlayabilirsiniz.

**Örnek Kod:** Aşağıdaki örnekte, etrafı çerçeveli
(`OutlineInputBorder`) ve ipucu metni içeren bir metin alanı
oluşturulmuştur.

``` dart
TextField(
  decoration: InputDecoration(
    border: OutlineInputBorder(), // Çerçeveli kenarlık
    hintText: 'Bir arama terimi girin', // İpucu metni
  ),
),
```

------------------------------------------------------------------------

## 2. TextFormField

`TextFormField`, aslında `TextField`’ı sarmalayan (wrap) ve onu kapsayan
bir **`Form`** widget’ı ile entegre eden bir yapıdır.

- **Kullanım Amacı:** Doğrulama (validation) işlemleri yapmak ve diğer
  `FormField` widget’ları ile birlikte çalışmak istediğinizde tercih
  edilir.
- **Form Entegrasyonu:** Formun durumu (state) üzerinden veriyi kaydetme
  veya doğrulama işlemlerini kolaylaştırır.

**Örnek Kod:** Aşağıdaki örnekte, altı çizili (`UnderlineInputBorder`)
ve etiketli bir form alanı oluşturulmuştur.

``` dart
TextFormField(
  decoration: const InputDecoration(
    border: UnderlineInputBorder(), // Altı çizili kenarlık
    labelText: 'Kullanıcı adınızı girin', // Etiket metni
  ),
),
```

------------------------------------------------------------------------

## Özet: Hangisini Seçmeli?

- Basit bir metin girişi istiyorsanız ve form doğrulama (validation)
  işlemlerine ihtiyacınız yoksa **`TextField`** kullanın.
- Bir kayıt formu veya giriş ekranı gibi, girilen verinin doğrulanması
  (örneğin: “Şifre en az 6 karakter olmalı”) gerekiyorsa
  **`TextFormField`** kullanın.

# Metin Alanından (TextField) Değer Okuma

Bu doküman, bir kullanıcının metin alanına girdiği veriyi
**`TextEditingController`** kullanarak nasıl alacağınızı ve
kullanacağınızı adım adım açıklar.

Kullanıcı girişini yakalamak için şu 3 adımı izleyeceğiz:

1.  Bir `TextEditingController` oluşturun.
2.  Bu controller’ı `TextField`’a bağlayın.
3.  Metin alanındaki güncel değeri okuyun.

------------------------------------------------------------------------

## 1. `TextEditingController` Oluşturma

Metin alanındaki veriyi yönetmek için öncelikle bir
`TextEditingController` nesnesi oluşturmalısınız. Bu işlem genellikle
bir `StatefulWidget` içinde yapılır.

> ⚠️ **Önemli:** `TextEditingController` ile işiniz bittiğinde
> (genellikle ekran kapandığında) **`dispose()`** metodunu çağırmayı
> unutmayın. Bu, nesnenin kullandığı kaynakların serbest bırakılmasını
> sağlar ve bellek sızıntılarını önler.

``` dart
// Özel bir Form widget'ı tanımlıyoruz
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

// Form ile ilgili verileri tutan State sınıfı
class _MyCustomFormState extends State<MyCustomForm> {
  // TextField'ın değerini almak için bir controller oluşturuyoruz
  final myController = TextEditingController();

  @override
  void dispose() {
    // Widget kapatıldığında controller'ı temizliyoruz (bellek yönetimi)
    myController.dispose();
    super.dispose();
  }

  @override
  Widget build(BuildContext context) {
    // Bir sonraki adımda burayı dolduracağız
    return Scaffold(
        // ...
    );
  }
}
```

------------------------------------------------------------------------

## 2. Controller’ı TextField’a Bağlama

Controller nesnesini oluşturduktan sonra, bunu `TextField` veya
`TextFormField` widget’ının **`controller`** özelliğine atamanız
gerekir. Bu işlem, controller’ın o metin alanını “dinlemesini” sağlar.

``` dart
return TextField(
  controller: myController, // Bağlantı burada yapılır
  decoration: InputDecoration(
    border: OutlineInputBorder(),
    hintText: 'Bir şeyler yazın...',
  ),
);
```

------------------------------------------------------------------------

## 3. Güncel Değeri Okuma ve Gösterme

Bağlantı sağlandıktan sonra, `myController.text` özelliğini kullanarak
kullanıcının girdiği metne (`String`) dilediğiniz zaman ulaşabilirsiniz.

Aşağıdaki örnekte, kullanıcı bir butona bastığında girilen metin bir
uyarı penceresinde (Alert Dialog) gösterilmektedir:

``` dart
FloatingActionButton(
  // Butona basıldığında metin alanındaki veriyi içeren bir diyalog göster
  onPressed: () {
    showDialog(
      context: context,
      builder: (context) {
        return AlertDialog(
          // TextEditingController kullanarak girilen metni alıyoruz
          content: Text(myController.text),
        );
      },
    );
  },
  tooltip: 'Değeri Göster!',
  child: const Icon(Icons.text_fields),
),
```

### Özet

- **Tanımla:** `final myController = TextEditingController();`
- **Bağla:** `TextField(controller: myController)`
- **Oku:** `print(myController.text);`
- **Temizle:** `myController.dispose();`

\## Örnek

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    return const MaterialApp(
      title: 'Retrieve Text Input',
      home: MyCustomForm(),
    );
  }
}

// Define a custom Form widget.
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

// Define a corresponding State class.
// This class holds the data related to the Form.
class _MyCustomFormState extends State<MyCustomForm> {
  // Create a text controller and use it to retrieve the current value
  // of the TextField.
  final myController = TextEditingController();

  @override
  void dispose() {
    // Clean up the controller when the widget is disposed.
    myController.dispose();
    super.dispose();
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text('Retrieve Text Input')),
      body: Padding(
        padding: const EdgeInsets.all(16),
        child: TextField(controller: myController),
      ),
      floatingActionButton: FloatingActionButton(
        // When the user presses the button, show an alert dialog containing
        // the text that the user has entered into the text field.
        onPressed: () {
          showDialog(
            context: context,
            builder: (context) {
              return AlertDialog(
                // Retrieve the text the that user has entered by using the
                // TextEditingController.
                content: Text(myController.text),
              );
            },
          );
        },
        tooltip: 'Show me the value!',
        child: const Icon(Icons.text_fields),
      ),
    );
  }
}
```

# Metin Alanı Değişikliklerini Yönetme

Bazı durumlarda, metin alanındaki yazı her değiştiğinde bir geri çağırma
(callback) fonksiyonu çalıştırmak isteyebilirsiniz. Örneğin, kullanıcı
yazdıkça sonuçları güncelleyen bir “otomatik tamamlama” özellikli arama
ekranı oluştururken buna ihtiyaç duyarsınız.

Flutter’da metin her değiştiğinde bir fonksiyonu tetiklemek için iki ana
yöntem vardır:

1.  `TextField` veya `TextFormField`’a bir **`onChanged()`** callback’i
    vermek.
2.  Bir **`TextEditingController`** kullanmak.

------------------------------------------------------------------------

## 1. `onChanged()` Callback’i Kullanma

En basit yaklaşım, `TextField` veya `TextFormField` widget’ına bir
`onChanged()` özelliği tanımlamaktır. Metin her değiştiğinde bu
fonksiyon otomatik olarak çağrılır.

Aşağıdaki örnekte, metin her değiştiğinde konsola mevcut değeri ve
karakter uzunluğunu yazdırıyoruz.

> **Önemli İpucu:** Kullanıcı girişiyle çalışırken `.length` yerine
> **`text.characters.length`** kullanmak önemlidir. Çünkü metin,
> emojiler veya karmaşık karakterler içerebilir. `characters` özelliği,
> kullanıcıya görünen karakter sayısının (grafem kümeleri) doğru
> sayılmasını sağlar.

``` dart
TextField(
  onChanged: (text) {
    print('İlk metin alanı: $text (${text.characters.length})');
  },
),
```

------------------------------------------------------------------------

## 2. `TextEditingController` Kullanma

Daha güçlü ama biraz daha ayrıntılı olan yaklaşım, `TextField` veya
`TextFormField`’ın **`controller`** özelliğine bir
`TextEditingController` atamaktır.

Metin değişikliklerinden haberdar olmak için şu adımları izleyerek
controller’ı dinlemelisiniz:

1.  Bir `TextEditingController` oluşturun.
2.  Controller’ı metin alanına bağlayın.
3.  En son değeri yazdıran bir fonksiyon oluşturun.
4.  Değişiklikleri yakalamak için controller’ı dinleyin (`addListener`).

### Adım 1: Controller Oluşturma

Bir `StatefulWidget` içerisinde controller’ı tanımlayın.

> **Not:** `TextEditingController` ile işiniz bittiğinde (genellikle
> widget ağacından kaldırıldığında) **`dispose()`** metodunu çağırmayı
> unutmayın. Bu, nesnenin kullandığı kaynakları serbest bırakır.

``` dart
// Özel Form widget'ı
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

class _MyCustomFormState extends State<MyCustomForm> {
  // TextField değerini almak için controller oluştur
  final myController = TextEditingController();

  @override
  void dispose() {
    // Widget kapatıldığında controller'ı temizle
    myController.dispose();
    super.dispose();
  }

  @override
  Widget build(BuildContext context) {
    // Bir sonraki adımda burası doldurulacak
    return const Placeholder(); 
  }
}
```

### Adım 2: Controller’ı Metin Alanına Bağlama

Oluşturduğunuz controller’ı `TextField`’a verin. Bu sayede iki sınıf
birbirine bağlanmış olur.

``` dart
TextField(controller: myController),
```

### Adım 3: Değer Okuyan Fonksiyonu Yazma

Metin her değiştiğinde çalışacak bir fonksiyona ihtiyacınız var.
`_MyCustomFormState` sınıfı içinde metnin o anki durumunu yazdıran bir
metot oluşturun.

``` dart
void _printLatestValue() {
  final text = myController.text;
  // Yine 'characters' özelliğini kullanarak doğru uzunluğu alıyoruz
  print('İkinci metin alanı: $text (${text.characters.length})');
}
```

### Adım 4: Değişiklikleri Dinleme (`addListener`)

Son olarak, `TextEditingController`’ı dinlemeye başlayın ve değişiklik
olduğunda `_printLatestValue()` metodunu çağırın. Bu işlem için
**`addListener()`** kullanılır.

Dinleme işlemini sınıf başlatıldığında (`initState`) başlatın ve sınıf
yok edildiğinde (`dispose`) sonlandırın.

``` dart
@override
void initState() {
  super.initState();

  // Değişiklikleri dinlemeye başla
  myController.addListener(_printLatestValue);
}

@override
void dispose() {
  // Widget ağaçtan kaldırıldığında controller'ı temizle.
  // Bu işlem, _printLatestValue dinleyicisini de otomatik olarak kaldırır.
  myController.dispose();
  super.dispose();
}
```

## İnteraktif Örnek

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    return const MaterialApp(
      title: 'Retrieve Text Input',
      home: MyCustomForm(),
    );
  }
}

// Define a custom Form widget.
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

// Define a corresponding State class.
// This class holds data related to the Form.
class _MyCustomFormState extends State<MyCustomForm> {
  // Create a text controller and use it to retrieve the current value
  // of the TextField.
  final myController = TextEditingController();

  @override
  void initState() {
    super.initState();

    // Start listening to changes.
    myController.addListener(_printLatestValue);
  }

  @override
  void dispose() {
    // Clean up the controller when the widget is removed from the widget tree.
    // This also removes the _printLatestValue listener.
    myController.dispose();
    super.dispose();
  }

  void _printLatestValue() {
    final text = myController.text;
    print('Second text field: $text (${text.characters.length})');
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text('Retrieve Text Input')),
      body: Padding(
        padding: const EdgeInsets.all(16),
        child: Column(
          children: [
            TextField(
              onChanged: (text) {
                print('First text field: $text (${text.characters.length})');
              },
            ),
            TextField(controller: myController),
          ],
        ),
      ),
    );
  }
}
```

# Flutter Ders Notları: Odak (Focus) ve Metin Alanları

Bir metin alanı seçildiğinde ve giriş kabul etmeye hazır olduğunda, o
alanın **“odaklandığı” (focus)** söylenir. Genellikle kullanıcılar bir
metin alanına dokunarak odağı kaydırır; ancak geliştiriciler bu işlemi,
bu rehberde açıklanan araçları kullanarak yazılımsal olarak da
yapabilirler.

Odak yönetimi, sezgisel ve akıcı formlar oluşturmak için temel bir
araçtır. Örneğin, bir arama ekranınız olduğunu varsayalım. Kullanıcı
arama ekranına geldiğinde, odağı otomatik olarak arama çubuğuna
ayarlayabilirsiniz. Böylece kullanıcı, metin alanına manuel olarak
dokunmak zorunda kalmadan, ekran açılır açılmaz yazmaya başlayabilir.

Bu dokümanda, bir metin alanı görünür görünmez ona nasıl
odaklanılacağını ve bir butona basıldığında odağın nasıl tetikleneceğini
öğreneceğiz.

------------------------------------------------------------------------

## 1. Görünür Görünmez Odaklanma (Autofocus)

Bir metin alanına, ekranda görünür görünmez odaklanmak için
**`autofocus`** özelliğini kullanın.

``` dart
TextField(
  autofocus: true, // Ekran açıldığında klavye otomatik açılır ve imleç bu alandadır
);
```

------------------------------------------------------------------------

## 2. Butona Basıldığında Odaklanma

Bazen odağı hemen değil, daha sonraki bir anda (örneğin bir API
çağrısından sonra veya bir doğrulama hatası oluştuğunda) belirli bir
metin alanına vermek isteyebilirsiniz.

Bu örnekte, kullanıcı bir butona bastığında odağı metin alanına şu
adımları izleyerek vereceğiz:

1.  Bir `FocusNode` oluşturun.
2.  `FocusNode`’u bir `TextField`’a atayın.
3.  Butona basıldığında `TextField`’a odaklanın.

### Adım 1: `FocusNode` Oluşturma

Öncelikle bir `FocusNode` oluşturun. `FocusNode`, Flutter’ın “odak
ağacı” (focus tree) içerisinde belirli bir `TextField`’ı tanımlamak için
kullanılır.

`FocusNode`’lar uzun ömürlü nesnelerdir, bu yüzden yaşam döngülerini
(lifecycle) bir `State` nesnesi içinde yönetmelisiniz. `initState`
metodunda örneği oluşturun ve `dispose` metodunda temizleyin.

``` dart
// Özel bir Form widget'ı tanımlıyoruz
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

// Form verilerini tutan State sınıfı
class _MyCustomFormState extends State<MyCustomForm> {
  // FocusNode'u tanımlayın. Yaşam döngüsünü yönetmek için;
  // initState içinde oluşturun, dispose içinde temizleyin.
  late FocusNode myFocusNode;

  @override
  void initState() {
    super.initState();
    myFocusNode = FocusNode();
  }

  @override
  void dispose() {
    // Form kapatıldığında node'u temizleyin (kaynak yönetimi)
    myFocusNode.dispose();
    super.dispose();
  }

  @override
  Widget build(BuildContext context) {
    // Sonraki adımda burayı dolduracağız.
    return Scaffold(
      // ...
    );
  }
}
```

### Adım 2: `FocusNode`’u `TextField`’a Atama

`FocusNode` nesnesini oluşturduktan sonra, bunu `build` metodu içindeki
ilgili `TextField`’ın `focusNode` özelliğine verin.

``` dart
@override
Widget build(BuildContext context) {
  return TextField(
    focusNode: myFocusNode, // Bağlantı burada kurulur
  );
}
```

### Adım 3: Butona Tıklandığında Odaklanma

Son olarak, kullanıcı bir butona (örneğin `FloatingActionButton`)
bastığında metin alanına odaklanmasını sağlayın. Bu işlem için
**`requestFocus()`** metodunu kullanın.

``` dart
FloatingActionButton(
  // Butona basıldığında, myFocusNode kullanarak metin alanına odaklan
  onPressed: () => myFocusNode.requestFocus(),
  child: const Icon(Icons.edit),
),
```

### Özet

- **Otomatik Odak:** `autofocus: true`
- **Manuel Odak:** `FocusNode` oluştur -\> `TextField`’a bağla -\>
  `myFocusNode.requestFocus()` çağır.

## İnteraktif Örnek

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    return const MaterialApp(title: 'Text Field Focus', home: MyCustomForm());
  }
}

// Define a custom Form widget.
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  State<MyCustomForm> createState() => _MyCustomFormState();
}

// Define a corresponding State class.
// This class holds data related to the form.
class _MyCustomFormState extends State<MyCustomForm> {
  // Define the focus node. To manage the lifecycle, create the FocusNode in
  // the initState method, and clean it up in the dispose method.
  late FocusNode myFocusNode;

  @override
  void initState() {
    super.initState();

    myFocusNode = FocusNode();
  }

  @override
  void dispose() {
    // Clean up the focus node when the Form is disposed.
    myFocusNode.dispose();

    super.dispose();
  }

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text('Text Field Focus')),
      body: Padding(
        padding: const EdgeInsets.all(16),
        child: Column(
          children: [
            // The first text field is focused on as soon as the app starts.
            const TextField(autofocus: true),
            // The second text field is focused on when a user taps the
            // FloatingActionButton.
            TextField(focusNode: myFocusNode),
          ],
        ),
      ),
      floatingActionButton: FloatingActionButton(
        // When the button is pressed,
        // give focus to the text field using myFocusNode.
        onPressed: () => myFocusNode.requestFocus(),
        tooltip: 'Focus Second Text Field',
        child: const Icon(Icons.edit),
      ), // This trailing comma makes auto-formatting nicer for build methods.
    );
  }
}
```

# Doğrulamalı Form Oluşturma

Uygulamalar genellikle kullanıcıların metin alanlarına bilgi girmesini
gerektirir (örneğin: giriş yapma, kayıt olma). Uygulamaları güvenli ve
kullanımı kolay hale getirmek için, kullanıcının sağladığı bilgilerin
geçerli olup olmadığını kontrol etmelisiniz.

Eğer kullanıcı formu doğru doldurduysa bilgiyi işlersiniz; yanlış
doldurduysa (örneğin boş bıraktıysa), neyin yanlış gittiğini bildiren
dostça bir hata mesajı gösterirsiniz.

Bu işlemi 3 temel adımda gerçekleştireceğiz:

1.  `GlobalKey` içeren bir `Form` oluşturmak.
2.  Doğrulama mantığına sahip bir `TextFormField` eklemek.
3.  Formu doğrulayıp gönderecek bir buton oluşturmak.

------------------------------------------------------------------------

## 1. `GlobalKey` ile Bir Form Oluşturma

`Form` widget’ı, birden fazla form alanını gruplamak ve doğrulamak için
bir kapsayıcı (container) görevi görür. Formu oluştururken ona benzersiz
bir kimlik atamak ve daha sonra doğrulama işlemini tetiklemek için bir
**`GlobalKey`** sağlamanız gerekir.

Bu işlem için `StatefulWidget` kullanılması önerilir, böylece anahtarı
(`key`) bir kez oluşturup saklayabilirsiniz.

> **Önemli:** `GlobalKey` oluşturmak kaynak açısından maliyetli bir
> işlemdir. Eğer `StatelessWidget` içinde her `build` metodunda yeniden
> anahtar üretirseniz, formun durumu sürekli sıfırlanır ve performans
> kaybı yaşanır.

``` dart
import 'package:flutter/material.dart';

// Özel Form widget'ı
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  MyCustomFormState createState() {
    return MyCustomFormState();
  }
}

class MyCustomFormState extends State<MyCustomForm> {
  // Form widget'ını benzersiz şekilde tanımlayan ve doğrulamaya izin veren
  // global bir anahtar oluşturuyoruz.
  // Not: Türü GlobalKey<FormState>'tir.
  final _formKey = GlobalKey<FormState>();

  @override
  Widget build(BuildContext context) {
    // Yukarıda oluşturulan _formKey'i kullanarak Form widget'ını oluşturun.
    return Form(
      key: _formKey,
      child: const Column(
        children: <Widget>[
          // TextFormField'lar ve Buton buraya gelecek.
        ],
      ),
    );
  }
}
```

------------------------------------------------------------------------

## 2. `TextFormField` ve Doğrulama Mantığı Ekleme

Kullanıcının metin girmesi için `TextFormField` kullanılır. Bu widget,
Material Design metin alanını oluşturur ve `Form` ile entegre çalışarak
doğrulama hatalarını görüntüleyebilir.

Doğrulama yapmak için **`validator()`** fonksiyonunu kullanırız:

- Kullanıcı metin girdiğinde bu fonksiyon çalışır.
- Girdi geçersizse: Bir **hata mesajı (String)** döndürür.
- Girdi geçerliyse: **`null`** döndürür.

Aşağıdaki örnekte, alanın boş olup olmadığını kontrol ediyoruz:

``` dart
TextFormField(
  // validator, kullanıcının girdiği metni (value) alır.
  validator: (value) {
    if (value == null || value.isEmpty) {
      return 'Lütfen bir metin giriniz'; // Hata mesajı
    }
    return null; // Sorun yok, geçerli
  },
),
```

------------------------------------------------------------------------

## 3. Doğrulama ve Gönderme Butonu

Son olarak, kullanıcının formu göndermesini sağlayacak bir buton
ekleyin. Butona basıldığında formun geçerli olup olmadığını kontrol
etmelisiniz.

Bunun için 1. adımda oluşturduğumuz `_formKey` kullanılır:

- `_formKey.currentState!.validate()` metodu çağrılır.
- Bu metod, form içindeki tüm `TextFormField`’ların `validator`
  fonksiyonlarını çalıştırır.
- Hepsi `null` dönerse (hata yoksa), metod `true` değerini verir.
- Herhangi biri hata mesajı dönerse, metod `false` verir ve form
  otomatik olarak güncellenerek hata mesajlarını ekranda gösterir.

``` dart
ElevatedButton(
  onPressed: () {
    // Form geçerli ise true, değilse false döner.
    if (_formKey.currentState!.validate()) {
      // Form geçerliyse, işlemi yapın (Veritabanına kayıt vb.)
      ScaffoldMessenger.of(context).showSnackBar(
        const SnackBar(content: Text('Veri İşleniyor...')),
      );
    }
  },
  child: const Text('Gönder'),
),
```

### Özet: Nasıl Çalışır?

`validate()` metodu çağrıldığında Flutter, formdaki her bir alanı tek
tek kontrol eder. Hatalı olan alanlar kırmızıya döner ve altında
yazdığınız hata mesajı belirir. Her şey doğruysa, `if` bloğunun içine
girilir ve işlem tamamlanır.

## İnteraktif Örnek

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    const appTitle = 'Form Validation Demo';

    return MaterialApp(
      title: appTitle,
      home: Scaffold(
        appBar: AppBar(title: const Text(appTitle)),
        body: const MyCustomForm(),
      ),
    );
  }
}

// Create a Form widget.
class MyCustomForm extends StatefulWidget {
  const MyCustomForm({super.key});

  @override
  MyCustomFormState createState() {
    return MyCustomFormState();
  }
}

// Create a corresponding State class.
// This class holds data related to the form.
class MyCustomFormState extends State<MyCustomForm> {
  // Create a global key that uniquely identifies the Form widget
  // and allows validation of the form.
  //
  // Note: This is a GlobalKey<FormState>,
  // not a GlobalKey<MyCustomFormState>.
  final _formKey = GlobalKey<FormState>();

  @override
  Widget build(BuildContext context) {
    // Build a Form widget using the _formKey created above.
    return Form(
      key: _formKey,
      child: Column(
        crossAxisAlignment: CrossAxisAlignment.start,
        children: [
          TextFormField(
            // The validator receives the text that the user has entered.
            validator: (value) {
              if (value == null || value.isEmpty) {
                return 'Please enter some text';
              }
              return null;
            },
          ),
          Padding(
            padding: const EdgeInsets.symmetric(vertical: 16),
            child: ElevatedButton(
              onPressed: () {
                // Validate returns true if the form is valid, or false otherwise.
                if (_formKey.currentState!.validate()) {
                  // If the form is valid, display a snackbar. In the real world,
                  // you'd often call a server or save the information in a database.
                  ScaffoldMessenger.of(context).showSnackBar(
                    const SnackBar(content: Text('Processing Data')),
                  );
                }
              },
              child: const Text('Submit'),
            ),
          ),
        ],
      ),
    );
  }
}
```

## 📄 Lisans Bilgisi

Bu doküman, **Flutter resmi dokümantasyonundan** türetilmiş Türkçe ders
notudur.

**Orijinal kaynak:**  
https://docs.flutter.dev/cookbook/forms/text-input

**Türkçe çeviri ve düzenleme:**  
[Doç. Dr. Hakan Temiz](mailto:htemiz@artvin.edu.tr)

------------------------------------------------------------------------

### Lisans Kapsamı

Bu dokümandaki içerikler aşağıdaki açık lisanslar kapsamında
sunulmaktadır:

**Metin içerikleri (anlatım ve açıklamalar):**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** Creative Commons Attribution 4.0 International (CC BY 4.0)  
https://creativecommons.org/licenses/by/4.0/

Bu lisans kapsamında: - İçerik kopyalanabilir, dağıtılabilir ve
uyarlanabilir  
- Ticari kullanım serbesttir  
- Kaynak belirtilmesi zorunludur

**Kod örnekleri:**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** BSD 3-Clause License  
https://opensource.org/licenses/BSD-3-Clause

Bu lisans kapsamında: - Kodlar kopyalanabilir, değiştirilebilir ve
dağıtılabilir  
- Ticari kullanım serbesttir  
- Lisans bildiriminin korunması gerekir
