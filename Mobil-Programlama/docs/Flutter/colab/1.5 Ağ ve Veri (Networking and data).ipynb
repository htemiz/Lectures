# Ağ ve Veri (Networking and data)

Her ne kadar “hiçbir insan bir ada değildir” denilse de, herhangi bir ağ
yeteneği olmayan bir Flutter uygulaması biraz kopuk hissettirebilir. Bu
sayfa, Flutter uygulamanıza ağ özelliklerini nasıl ekleyeceğinizi
kapsar. Uygulamanız veri alacak, JSON’u bellekte kullanılabilir
temsillere ayrıştıracak ve ardından verileri tekrar dışarı
gönderecektir.

## Ağ üzerinden veri almaya giriş

En basit haliyle, Dart VM tabanlı platformlar ve web tarayıcısı tabanlı
ortamlar arasındaki ağ erişimi farklarına uyum sağlamak için `http`
paketini kullandığınızı varsayarsak, bir HTTP `GET` isteği yapmak
aşağıdaki kadar basit olabilir:

``` dart
import 'package:http/http.dart' as http;

void main() async {
  var response = await http.get(
    Uri.parse('https://jsonplaceholder.typicode.com/albums/1'),
  );
  print(response.body);
}
```

Aşağıdaki iki eğitim, ister Android, iOS, bir web tarayıcısı içinde,
ister Windows, macOS veya Linux üzerinde yerel olarak çalışıyor olun,
uygulamanıza `http` paketini eklemeyle ilgili tüm ayrıntıları gösterir.
İlk eğitim, bir web sitesine kimlik doğrulaması gerektirmeyen bir `GET`
isteği yapmayı, alınan verileri `JSON` olarak ayrıştırmayı ve ardından
sonuçtaki verileri görüntülemeyi gösterir. İkinci eğitim, kimlik
doğrulama başlıkları ekleyerek ve yetkilendirme gerektiren web
sunucularına erişimi etkinleştirerek ilkinin üzerine inşa eder. Mozilla
Geliştirici Ağı (MDN) makalesi, web’de yetkilendirmenin nasıl
çalıştığına dair daha fazla arka plan bilgisi verir.

- **Eğitim:** [İnternetten veri
  çekme](https://docs.flutter.dev/cookbook/networking/fetch-data)
- **Eğitim:** [Kimlik doğrulamalı istekler
  yapma](https://docs.flutter.dev/cookbook/networking/authenticated-requests)
- **Makale:** [Web siteleri için Yetkilendirme üzerine MDN
  makalesi](https://developer.mozilla.org/en-US/docs/Web/HTTP/Authentication)

## Ağdan alınan verileri kullanışlı hale getirme

Ağdan veri aldığınızda, verileri ağdan Dart’ta kolayca çalışabileceğiniz
bir şeye dönüştürmenin bir yoluna ihtiyacınız vardır. Önceki bölümdeki
eğitimler, ağ verilerini bellek içi bir temsile dönüştürmek için elle
yazılmış Dart kodu kullandı.

Bu bölümde, bu dönüşümü gerçekleştirmenin diğer seçeneklerini
göreceksiniz. İlki, `freezed` paketine genel bir bakış sunan bir YouTube
videosuna bağlantı verir. İkincisi, JSON ayrıştırma vaka çalışmasını
kullanarak kalıpları (patterns) ve kayıtları (records) kapsayan bir
codelab’e bağlantı verir.

- **YouTube videosu:** [Freezed (Haftanın
  Paketi)](https://www.google.com/search?q=https://www.youtube.com/watch%3Fv%3DRaThk0fWCFQ)
- **Codelab:** [Dart’ın kalıplarına ve kayıtlarına
  dalış](https://codelabs.developers.google.com/codelabs/dart-patterns-records)

## Her iki yöne gitmek, verileri tekrar dışarı çıkarmak

Artık veri alma sanatında ustalaştığınıza göre, verileri dışarı itmeye
bakmanın zamanı geldi. Bu bilgi verileri ağa göndermekle başlar, ancak
daha sonra asenkronluğa dalar. Gerçek şu ki, ağ üzerinden bir sohbete
girdiğinizde, fiziksel olarak uzaktaki web sunucularının yanıt
vermesinin biraz zaman alabileceği gerçeğiyle başa çıkmanız gerekecek ve
paketlerin gidiş-dönüş yapmasını beklerken ekrana görüntü çizmeyi
durduramazsınız.

Dart’ın asenkronluk için harika bir desteği vardır, tıpkı Flutter gibi.
Dart’ın desteği hakkında her şeyi bir eğitimde öğrenecek, ardından
Flutter’ın yeteneğinin Haftanın Widget’ı videosunda ele alındığını
göreceksiniz. Bunu tamamladığınızda, DevTool’un Ağ Görünümü’nü
kullanarak ağ trafiğinde nasıl hata ayıklayacağınızı öğreneceksiniz.

- **Eğitim:** [İnternete veri
  gönderme](https://docs.flutter.dev/cookbook/networking/send-data)
- **Eğitim:** [Asenkron programlama: futures, async,
  await](https://dart.dev/codelabs/async-await)
- **YouTube videosu:** [FutureBuilder (Haftanın
  Widget’ı)](https://www.youtube.com/watch?v=ek8ZPdWj4Qo)
- **Makale:** [Ağ Görünümü’nü
  Kullanma](https://docs.flutter.dev/tools/devtools/network)

## Ek materyal

Artık Flutter’ın ağ API’lerini kullanmada ustalaştığınıza göre,
Flutter’ın ağ kullanımını bağlam içinde görmek faydalı olacaktır. İlk
codelab (görünüşte Flutter’da Uyarlanabilir uygulamalar oluşturma
üzerine), web tarayıcılarının Cross-Origin Resource Sharing (CORS)
kısıtlamalarını aşmak için Dart ile yazılmış bir web sunucusu kullanır.

> **Not** Eğer bu codelab’i [düzen](https://docs.flutter.dev/ui/layout)
> sayfasında zaten tamamladıysanız, bu adımı atlamaktan çekinmeyin.

Ardından, Flutter DevRel mezunu Fitz’in, Flutter uygulamaları için
verilerin konumunun ne kadar önemli olduğu hakkında konuştuğu uzun
soluklu bir YouTube videosu. Son olarak, Flutter GDE Anna (Domashych)
Leushchenko tarafından Flutter’da gelişmiş ağ işlemlerini kapsayan
gerçekten yararlı bir makale serisi.

- **Codelab:** [Flutter’da uyarlanabilir
  uygulamalar](https://codelabs.developers.google.com/codelabs/flutter-adaptive-app)
- **Video:** [Yerelde tutmak: Bir Flutter uygulamasının verilerini
  yönetmek](https://www.google.com/search?q=https://www.youtube.com/watch%3Fv%3D7gQnSjQ87ks)
- **Makale serisi:** [Dart ve Flutter’da temel ve gelişmiş ağ
  işlemleri](https://www.google.com/search?q=https://medium.com/flutter-community/basic-and-advanced-networking-in-dart-and-flutter-the-tide-is-high-but-im-holding-on-7c26df502287)

## 📄 Lisans Bilgisi

Bu doküman, **Flutter resmi dokümantasyonundan** türetilmiş Türkçe ders
notudur.

**Orijinal kaynak:**  
https://docs.flutter.dev/get-started/fundamentals/networking

**Türkçe çeviri ve düzenleme:**  
[Doç. Dr. Hakan Temiz](mailto:htemiz@artvin.edu.tr)

------------------------------------------------------------------------

### Lisans Kapsamı

Bu dokümandaki içerikler aşağıdaki açık lisanslar kapsamında
sunulmaktadır:

**Metin içerikleri (anlatım ve açıklamalar):**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** Creative Commons Attribution 4.0 International (CC BY 4.0)  
https://creativecommons.org/licenses/by/4.0/

Bu lisans kapsamında: - İçerik kopyalanabilir, dağıtılabilir ve
uyarlanabilir  
- Ticari kullanım serbesttir  
- Kaynak belirtilmesi zorunludur

**Kod örnekleri:**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** BSD 3-Clause License  
https://opensource.org/licenses/BSD-3-Clause

Bu lisans kapsamında: - Kodlar kopyalanabilir, değiştirilebilir ve
dağıtılabilir  
- Ticari kullanım serbesttir  
- Lisans bildiriminin korunması gerekir
