# Kaydırma (Scrolling)

## Flutter’ın kaydırma desteğine genel bakış

Flutter, otomatik olarak kayan birçok yerleşik widget’a sahiptir ve
ayrıca belirli kaydırma davranışları oluşturmak için
özelleştirebileceğiniz çeşitli widget’lar sunar.

### Temel kaydırma

Birçok Flutter widget’ı, kutudan çıktığı gibi (hazır olarak) kaydırmayı
destekler ve işin çoğunu sizin için yapar. Örneğin,
`SingleChildScrollView` gerektiğinde çocuğunu otomatik olarak kaydırır.
Diğer yararlı widget’lar arasında `ListView` ve `GridView` bulunur. Bu
widget’ların daha fazlasını Widget kataloğunun **kaydırma sayfasında**
inceleyebilirsiniz.

[Video](https://www.youtube.com/watch?v=DbkIQSvwnZc)

[Video](https://www.youtube.com/watch?v=KJpkjHGiI5A)

### Sonsuz kaydırma

`ListView` veya `GridView` içinde uzun bir öğe listeniz olduğunda
(sonsuz bir liste dahil), öğeleri görünüme girdikçe talep üzerine
oluşturabilirsiniz. Bu, çok daha performanslı bir kaydırma deneyimi
sağlar. Daha fazla bilgi için `ListView.builder` veya
`GridView.builder`’a göz atın.

### Gelişmiş (Fancy) kaydırma

Belki de kaydırma sıçraması (scroll bouncing) olarak da adlandırılan
elastik kaydırmayı uygulamak istiyorsunuz. Veya belki paralaks kaydırma
gibi diğer dinamik kaydırma efektlerini uygulamak istiyorsunuz. Ya da
küçülme veya kaybolma gibi çok spesifik davranışlara sahip kayan bir
başlık istiyorsunuz.

Tüm bunları ve daha fazlasını Flutter `Sliver*` sınıflarını kullanarak
başarabilirsiniz. Bir sliver (dilim), kaydırılabilir alanın bir
parçasını ifade eder. O alan üzerinde daha ince ayarlı bir kontrole
sahip olmak için bir sliver tanımlayabilir ve bunu bir
`CustomScrollView` içine ekleyebilirsiniz.

Daha fazla bilgi için **Süslü kaydırma elde etmek için sliver’ları
kullanma** ve `Sliver` sınıflarına göz atın.
