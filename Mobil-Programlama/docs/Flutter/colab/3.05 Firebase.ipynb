# Firebase

Firebase, gerçek zamanlı veritabanı (realtime database), bulut depolama
(cloud storage), kimlik doğrulama (authentication), kilitlenme raporlama
(crash reporting), makine öğrenimi, uzaktan yapılandırma (remote
configuration) ve statik dosyalarınız için barındırma (hosting) gibi
barındırılan arka uç hizmetleri sağlayan bir Hizmet Olarak Arka Uç
(BaaS) uygulama geliştirme platformudur.

# Firebase’i Flutter Uygulamanıza Ekleyin

Bu kılavuz, Flutter uygulamanızda Firebase hizmetlerini nasıl kullanmaya
başlayacağınızı anlatır. Hem iOS, Android hem de Web platformlarını
destekler.

## Ön Koşullar

Başlamadan önce aşağıdakilere sahip olduğunuzdan emin olun:

- **Editör:** Tercih ettiğiniz düzenleyici veya IDE (VS Code, Android
  Studio vb.) yüklü olmalıdır.

- **Cihaz/Emülatör:** Uygulamanızı çalıştırmak için bir cihaz veya
  emülatör.

- *Emülatör Notu:* Emülatörler Google Play’i içeren bir sistem görüntüsü
  (system image) kullanmalıdır.

- **Android Gereksinimleri:**

- API düzeyi 23 (Marshmallow) veya sonraki sürümleri hedeflemelidir.

- Cihazda Android 6.0 veya sonraki bir sürüm yüklü olmalıdır.

- **Flutter Kurulumu:** İşletim sisteminiz için şunlar yüklenmiş
  olmalıdır:

- Flutter SDK

- Desteklenen kütüphaneler

- Platforma özel yazılımlar ve SDK’lar

- **Firebase Hesabı:** Google Hesabınızı kullanarak Firebase’de oturum
  açmış olmalısınız.

> **Not:** Henüz bir uygulamanız yoksa, `flutter create` komutuyla veya
> IDE’niz aracılığıyla yeni bir proje oluşturun.

------------------------------------------------------------------------

## 1. Adım: Gerekli komut satırı araçlarını yükleyin

1.  Henüz yapmadıysanız **Firebase CLI**’yı yükleyin.
2.  Aşağıdaki komutu çalıştırarak Google Hesabınızla Firebase’e giriş
    yapın:

``` bash
firebase login
```

1.  **FlutterFire CLI**’yı yüklemek için herhangi bir dizinden şu komutu
    çalıştırın:

``` bash
dart pub global activate flutterfire_cli
```

## 2. Adım: Uygulamanızı yapılandırın

Flutter uygulamalarınızı Firebase’e bağlamak için **FlutterFire CLI**
kullanılır. Bu araç, Firebase projenizle Flutter projeniz arasındaki
köprüyü kurar.

1.  Uygulama yapılandırma iş akışını başlatmak için **Flutter proje
    dizininizden** aşağıdaki komutu çalıştırın:

``` bash
flutterfire configure
```

### `flutterfire configure` ne işe yarar?

Bu komut, projeniz için bir `firebase_options.dart` dosyası oluşturur ve
uygulamanızı Firebase projenize kaydeder. Aşağıdaki durumlarda bu komutu
**yeniden çalıştırmanız** gerekir:

- Flutter uygulamanızda yeni bir platformu (örneğin Web veya MacOS)
  desteklemeye başladığınızda.
- Yeni bir Firebase hizmeti kullanmaya başladığınızda (özellikle Google
  Sign-In, Crashlytics, Performance Monitoring veya Realtime Database).

------------------------------------------------------------------------

## 3. Adım: Uygulamanızda Firebase’i başlatın

1.  Temel Firebase çekirdek eklentisini (`firebase_core`) yüklemek için
    proje dizininde şu komutu çalıştırın:

``` bash
flutter pub add firebase_core
```

1.  Yapılandırmanın güncel olduğundan emin olmak için tekrar
    yapılandırma komutunu çalıştırın:

``` bash
flutterfire configure
```

1.  `lib/main.dart` dosyanızda gerekli paketleri içe aktarın:

``` dart
import 'package:firebase_core/firebase_core.dart';
import 'firebase_options.dart';
```

1.  Yine `lib/main.dart` dosyanızda, `main` fonksiyonunu `async` hale
    getirin ve Firebase’i başlatın:

``` dart
void main() async {
  WidgetsFlutterBinding.ensureInitialized();
  await Firebase.initializeApp(
    options: DefaultFirebaseOptions.currentPlatform,
  );
  runApp(const MyApp());
}
```

1.  Uygulamanızı çalıştırın:

``` bash
flutter run
```

> **Demo Projesi İpucu:** Eğer Firebase Emülatörlerini kullanıyorsanız
> veya bir demo projesiyle çalışıyorsanız, başlatma kodunu şu şekilde
> `demoProjectId` ile yapabilirsiniz:
>
> ``` dart
> await Firebase.initializeApp(
>   demoProjectId: "demo-project-id",
> );
> ```

------------------------------------------------------------------------

## 4. Adım: Firebase eklentilerini ekleyin

Firebase ürünlerine (Authentication, Firestore vb.) erişmek için ilgili
eklentileri yüklemeniz gerekir.

1.  İstediğiniz eklentiyi ekleyin (Örneğin: `firebase_auth`):

``` bash
flutter pub add PLUGIN_NAME
```

1.  **Önemli:** Her yeni eklenti eklediğinizde, Android/iOS tarafındaki
    yerel yapılandırmaların (Gradle eklentileri vb.) güncellenmesi için
    yapılandırma komutunu tekrar çalıştırın:

``` bash
flutterfire configure
```

1.  Uygulamayı yeniden derleyin:

``` bash
flutter run
```

### Mevcut Eklentiler

| Ürün | Eklenti Adı | Destek Durumu |
|------------------------|------------------------|------------------------|
| **Firebase AI Logic** | `firebase_vertexai` | Beta (Eski adıyla `firebase_ai`) |
| **Analytics** | `firebase_analytics` | Beta |
| **App Check** | `firebase_app_check` | Beta |
| **Authentication** | `firebase_auth` | Beta |
| **Cloud Firestore** | `cloud_firestore` | Beta |
| **Cloud Functions** | `cloud_functions` | Beta |
| **Cloud Messaging** | `firebase_messaging` | Beta |
| **Cloud Storage** | `firebase_storage` | Beta |
| **Crashlytics** | `firebase_crashlytics` | Beta |
| **Data Connect** | `firebase_data_connect` | \- |
| **Dynamic Links** | `firebase_dynamic_links` | \- |
| **In-App Messaging** | `firebase_in_app_messaging` | \- |
| **Installations** | `firebase_app_installations` | Beta |
| **ML Model Downloader** | `firebase_ml_model_downloader` | Beta |
| **Performance** | `firebase_performance` | \- |
| **Realtime Database** | `firebase_database` | Beta |
| **Remote Config** | `firebase_remote_config` | Beta |

> **Dikkat:** Windows üzerinde Firebase desteği şu an için üretim
> (production) amaçlı değildir, yalnızca yerel geliştirme iş akışları
> içindir.

------------------------------------------------------------------------

## Örnek: Analytics ile Uygulamayı Test Etme

Kurulumunuzun doğru yapıldığını test etmek için basit bir Analytics
entegrasyonu yapabilirsiniz.

1.  Firebase yapılandırmasını tamamladığınız bir projeyi açın.
2.  `lib/main.dart` dosyasının içeriğini silin.
3.  [Google Analytics Örnek
    Deposu](https://github.com/firebase/flutterfire/tree/master/packages/firebase_analytics/firebase_analytics/example)’ndan
    `main.dart` ve `tabs_page.dart` kodlarını alıp projenize yapıştırın.
4.  Uygulamayı çalıştırın (`flutter run`).
5.  **Firebase Konsolu**’na gidin ve sol menüden **Analytics** \>
    **Dashboard** (Kontrol Paneli) seçeneğine tıklayın.
6.  “Son 30 dakika içinde etkin olan kullanıcı sayısı” panelini kontrol
    edin. Ayrıca anlık olayları görmek için **DebugView** sekmesini
    kullanabilirsiniz.

## 📄 Lisans Bilgisi

Bu doküman, **Flutter resmi dokümantasyonundan** türetilmiş Türkçe ders
notudur.

**Orijinal kaynak:**  
https://docs.flutter.dev/data-and-backend/firebase

**Türkçe çeviri ve düzenleme:**  
[Doç. Dr. Hakan Temiz](mailto:htemiz@artvin.edu.tr)

------------------------------------------------------------------------

### Lisans Kapsamı

Bu dokümandaki içerikler aşağıdaki açık lisanslar kapsamında
sunulmaktadır:

**Metin içerikleri (anlatım ve açıklamalar):**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** Creative Commons Attribution 4.0 International (CC BY 4.0)  
https://creativecommons.org/licenses/by/4.0/

Bu lisans kapsamında: - İçerik kopyalanabilir, dağıtılabilir ve
uyarlanabilir  
- Ticari kullanım serbesttir  
- Kaynak belirtilmesi zorunludur

**Kod örnekleri:**  
Flutter resmi dokümantasyonundan alınmış veya uyarlanmıştır.  
**Lisans:** BSD 3-Clause License  
https://opensource.org/licenses/BSD-3-Clause

Bu lisans kapsamında: - Kodlar kopyalanabilir, değiştirilebilir ve
dağıtılabilir  
- Ticari kullanım serbesttir  
- Lisans bildiriminin korunması gerekir
