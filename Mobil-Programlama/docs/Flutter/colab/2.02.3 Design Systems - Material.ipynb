# Material Bileşenleri Widget’ları (Material Components Widgets)

Material Design, dijital deneyimler oluşturmak için esnek bir kılavuz
sistemidir. Flutter, Material Design’ı uygulayan zengin bir widget seti
sunar. Bu widget’lar, uygulamanızda Material Design görünümünü ve
hissini (look and feel) kolayca elde etmenizi sağlar.

Aşağıda, işlevlerine göre kategorize edilmiş temel Material widget’ları
bulunmaktadır.

## Uygulama yapısı ve gezinme (App structure and navigation)

Bu widget’lar uygulamanızın temel yapısını oluşturur ve ekranlar arası
gezinmeyi sağlar.

- **[App
  bar](https://api.flutter.dev/flutter/material/AppBar-class.html):**
  Ekranın üst kısmında bulunan, araç çubuğu ve diğer etkileşimli öğeleri
  içeren bir çubuktur. Başlık, eylemler ve gezinme kontrollerini
  barındırır.
- **[Bottom app
  bar](https://api.flutter.dev/flutter/material/BottomAppBar-class.html):**
  Ekranın altında gezinme ve eylemler sunan bir çubuktur. Genellikle bir
  `FloatingActionButton` ile birlikte kullanılır.
- **[Bottom navigation
  bar](https://api.flutter.dev/flutter/material/BottomNavigationBar-class.html):**
  Uygulamanın birincil hedefleri arasında geçiş yapmayı sağlayan alt
  gezinme çubuğudur.
- **[Drawer](https://api.flutter.dev/flutter/material/Drawer-class.html):**
  Ekranın kenarından (genellikle soldan) kayarak açılan bir Material
  tasarım panelidir. Uygulamanın gezinme bağlantılarını içerir.
- **[MaterialApp](https://api.flutter.dev/flutter/material/MaterialApp-class.html):**
  Material Design kullanan uygulamalar için gerekli olan tema, gezinme
  ve yerelleştirme gibi birçok özelliği yapılandıran kök widget’tır.
- **[Navigation
  bar](https://api.flutter.dev/flutter/material/NavigationBar-class.html):**
  Material 3 tasarımına uygun, uygulamanın ana bölümleri arasında geçiş
  yapmayı sağlayan modern alt gezinme çubuğudur.
- **[Navigation
  drawer](https://api.flutter.dev/flutter/material/NavigationDrawer-class.html):**
  Material 3 tasarımına uygun, ekranın yanından açılan gezinme paneli.
- **[Navigation
  rail](https://api.flutter.dev/flutter/material/NavigationRail-class.html):**
  Geniş ekranlı cihazlarda (tablet, masaüstü) ekranın sol veya sağ
  kenarında dikey olarak görüntülenen gezinme çubuğudur.
- **[Scaffold](https://api.flutter.dev/flutter/material/Scaffold-class.html):**
  Temel Material Design görsel düzen yapısını uygular. App bar, drawer,
  snack bar ve bottom sheet gibi yaygın öğeler için API sağlar.
- **[TabBar](https://api.flutter.dev/flutter/material/TabBar-class.html):**
  Yatay bir sekme düzeni görüntüler.
- **[TabBarView](https://api.flutter.dev/flutter/material/TabBarView-class.html):**
  Şu anda seçili olan sekmeye karşılık gelen widget’ı görüntüler.

## Düğmeler (Buttons)

Kullanıcıların eylem gerçekleştirmesini sağlayan çeşitli düğme türleri.

- **[ElevatedButton](https://api.flutter.dev/flutter/material/ElevatedButton-class.html):**
  Yükseltilmiş (gölgeli) bir düğme. Önemli eylemleri vurgulamak için
  kullanılır.
- **[FilledButton](https://api.flutter.dev/flutter/material/FilledButton-class.html):**
  Dolgulu ve gölgesiz düğme. Önemli eylemler için kullanılır.
- **[FloatingActionButton](https://api.flutter.dev/flutter/material/FloatingActionButton-class.html):**
  Ekranın içeriğinin üzerinde “yüzen” dairesel bir simge düğmesidir.
  Uygulamadaki birincil eylemi temsil eder.
- **[IconButton](https://api.flutter.dev/flutter/material/IconButton-class.html):**
  Bir simgenin içine basıldığında mürekkep (ink) tepkisi veren bir
  düğmedir.
- **[OutlinedButton](https://api.flutter.dev/flutter/material/OutlinedButton-class.html):**
  Kenarlıklı ancak dolgusuz bir düğme. İkincil eylemler için kullanılır.
- **[SegmentedButton](https://api.flutter.dev/flutter/material/SegmentedButton-class.html):**
  Seçenekleri, görünümleri değiştirmek veya öğeleri sıralamak için
  kullanılan bölümlere ayrılmış düğme grubu.
- **[TextButton](https://api.flutter.dev/flutter/material/TextButton-class.html):**
  Görünür bir kenarlığı veya arka planı olmayan, sadece metin (veya
  simge) içeren düğme. Diyaloglarda veya kartlarda yaygın olarak
  kullanılır.

## Giriş ve seçimler (Input and selections)

Kullanıcıdan veri almak veya seçim yapmasını sağlamak için kullanılan
widget’lar.

- **[Checkbox](https://api.flutter.dev/flutter/material/Checkbox-class.html):**
  Kullanıcının bir listeden birden fazla öğe seçmesine izin veren onay
  kutusu.
- **[Date & Time
  Pickers](https://api.flutter.dev/flutter/material/showDatePicker.html):**
  Kullanıcının tarih veya saat seçmesini sağlayan diyaloglar
  (`showDatePicker`, `showTimePicker`).
- **[Radio](https://api.flutter.dev/flutter/material/Radio-class.html):**
  Kullanıcının bir grup seçenek arasından yalnızca birini seçmesine izin
  veren radyo düğmesi.
- **[Slider](https://api.flutter.dev/flutter/material/Slider-class.html):**
  Kullanıcının bir değer aralığından (örneğin ses seviyesi, parlaklık)
  bir değer seçmesini sağlayan kaydırıcı.
- **[Switch](https://api.flutter.dev/flutter/material/Switch-class.html):**
  Açık/kapalı durumunu değiştiren bir anahtar.
- **[TextField](https://api.flutter.dev/flutter/material/TextField-class.html):**
  Kullanıcının metin girmesini sağlayan bir alan.

## İletişim kutuları, uyarılar ve paneller (Dialogs, alerts, and panels)

Kullanıcıya bilgi vermek veya onay almak için kullanılan bileşenler.

- **[AlertDialog](https://api.flutter.dev/flutter/material/AlertDialog-class.html):**
  Kullanıcıyı bir durumu kabul etmesi veya bir bilgi hakkında
  bilgilendirmesi gereken durumlarda ortaya çıkan bir iletişim kutusu.
- **[BottomSheet](https://api.flutter.dev/flutter/material/BottomSheet-class.html):**
  Ekranın altından kayarak gelen ek bir içerik panelidir. Kalıcı
  (Persistent) veya kalıcı olmayan (Modal) olabilir.
- **[ExpansionPanel](https://api.flutter.dev/flutter/material/ExpansionPanel-class.html):**
  Başlığına dokunulduğunda ayrıntıları göstermek için genişleyen veya
  daralan bir panel.
- **[SimpleDialog](https://api.flutter.dev/flutter/material/SimpleDialog-class.html):**
  Kullanıcıya birden fazla seçenek sunan basit bir iletişim kutusu.
- **[SnackBar](https://api.flutter.dev/flutter/material/SnackBar-class.html):**
  Ekranın alt kısmında kısa süreliğine görünen ve bir işlem hakkında
  geri bildirim sağlayan hafif bir mesaj çubuğu.

## Bilgi ekranları (Information displays)

Bilgileri düzenli ve görsel olarak çekici bir şekilde sunan widget’lar.

- **[Badge](https://api.flutter.dev/flutter/material/Badge-class.html):**
  Sayı veya durum gibi dinamik bilgileri iletmek için simgelerin
  üzerinde kullanılan küçük işaretçiler.
- **[Card](https://api.flutter.dev/flutter/material/Card-class.html):**
  İlgili bilgileri (albüm, coğrafi konum, yemek, iletişim bilgileri vb.)
  yuvarlatılmış köşelere ve gölgeye sahip bir kutuda gruplayan bir
  paneldir.
- **[Chip](https://api.flutter.dev/flutter/material/Chip-class.html):**
  Karmaşık bir varlığı (örneğin bir kişi, yer veya eylem) temsil eden
  kompakt bir öğedir. Giriş, seçim veya eylem için kullanılabilir.
- **[CircularProgressIndicator](https://api.flutter.dev/flutter/material/CircularProgressIndicator-class.html):**
  Bir işlemin devam ettiğini gösteren dairesel, dönen bir gösterge.
- **[DataTable](https://api.flutter.dev/flutter/material/DataTable-class.html):**
  Verileri satırlar ve sütunlar halinde gösteren bir tablo.
- **[GridView](https://api.flutter.dev/flutter/widgets/GridView-class.html):**
  Öğeleri iki boyutlu bir ızgara düzeninde görüntüler.
- **[Icon](https://api.flutter.dev/flutter/widgets/Icon-class.html):**
  Material Design simgelerini (veya diğer simge setlerini) görüntüler.
- **[Image](https://api.flutter.dev/flutter/widgets/Image-class.html):**
  Bir resim görüntüler.
- **[LinearProgressIndicator](https://api.flutter.dev/flutter/material/LinearProgressIndicator-class.html):**
  Bir işlemin ilerleyişini gösteren yatay çubuk.
- **[ListTile](https://api.flutter.dev/flutter/material/ListTile-class.html):**
  Genellikle bir simge, başlık ve alt başlıktan oluşan tek bir sabit
  yükseklikteki satır. Listelerde yaygın olarak kullanılır.
- **[Tooltip](https://api.flutter.dev/flutter/material/Tooltip-class.html):**
  Bir widget’ın üzerine gelindiğinde veya uzun basıldığında açıklayıcı
  bir metin etiketi gösterir.

## Düzen (Layout)

İçeriği düzenlemek ve ayırmak için kullanılan yardımcı widget’lar.

- **[Divider](https://api.flutter.dev/flutter/material/Divider-class.html):**
  İçerikleri listelerde veya sayfa düzenlerinde ayırmak için kullanılan
  ince bir çizgi.
- **[Stepper](https://api.flutter.dev/flutter/material/Stepper-class.html):**
  İlerlemeyi bir dizi adım aracılığıyla görüntüleyen bir Material
  widget’ı.
