# Cupertino (iOS tarzı) Widget’lar

Flutter’ın iOS tarzı widget’ları, Apple’ın iOS ve macOS için İnsan
Arayüzü Yönergeleri (Human Interface Guidelines) ile uyumlu, güzel ve
yüksek kaliteli bir kullanıcı arayüzü oluşturmanızı sağlar.

Aşağıda, işlevlerine göre kategorize edilmiş temel Cupertino widget’ları
bulunmaktadır.

## Uygulama yapısı ve gezinme (App structure and navigation)

Bu widget’lar, iOS tarzı bir uygulamanın temel yapısını ve gezinme
öğelerini oluşturur.

- **[CupertinoApp](https://api.flutter.dev/flutter/cupertino/CupertinoApp-class.html):**
  Cupertino (iOS) tasarımını kullanan uygulamalar için gerekli temayı ve
  yapılandırmayı sağlayan kök widget’tır.
- **[CupertinoPageScaffold](https://api.flutter.dev/flutter/cupertino/CupertinoPageScaffold-class.html):**
  Temel iOS tarzı sayfa düzeni yapısıdır. Bir gezinme çubuğunu ve
  içeriği bir arka plan üzerinde konumlandırır.
- **[CupertinoTabScaffold](https://api.flutter.dev/flutter/cupertino/CupertinoTabScaffold-class.html):**
  Sekmeli iOS uygulama yapısıdır. İçerik sekmelerinin üzerine bir sekme
  çubuğu (tab bar) yerleştirir.
- **[CupertinoNavigationBar](https://api.flutter.dev/flutter/cupertino/CupertinoNavigationBar-class.html):**
  Ekranın üst kısmında bulunan iOS tarzı gezinme çubuğudur. Genellikle
  `CupertinoPageScaffold` ile kullanılır.
- **[CupertinoSliverNavigationBar](https://api.flutter.dev/flutter/cupertino/CupertinoSliverNavigationBar-class.html):**
  iOS 11 tarzı büyük başlıkları (large titles) destekleyen
  kaydırılabilir (sliver) gezinme çubuğu.
- **[CupertinoTabBar](https://api.flutter.dev/flutter/cupertino/CupertinoTabBar-class.html):**
  iOS tarzı alt sekme çubuğudur. Genellikle `CupertinoTabScaffold` ile
  kullanılır.
- **[CupertinoTabView](https://api.flutter.dev/flutter/cupertino/CupertinoTabView-class.html):**
  Sekmeler arasında paralel gezinmeyi destekleyen bir sekmenin kök
  içeriğidir.
- **[CupertinoPageTransition](https://api.flutter.dev/flutter/cupertino/CupertinoPageTransition-class.html):**
  iOS tarzı sayfa geçiş animasyonu sağlar.

## Düğmeler (Buttons)

Kullanıcı etkileşimi için çeşitli düğme türleri.

- **[CupertinoButton](https://api.flutter.dev/flutter/cupertino/CupertinoButton-class.html):**
  iOS tarzı bir düğmedir.
- **[CupertinoContextMenuAction](https://api.flutter.dev/flutter/cupertino/CupertinoContextMenuAction-class.html):**
  Bağlam menüsü (ContextMenu) içinde kullanılan bir düğme.

## Giriş ve seçimler (Input and selections)

Kullanıcıdan veri almak veya seçim yapmasını sağlamak için kullanılan
widget’lar.

- **[CupertinoTextField](https://api.flutter.dev/flutter/cupertino/CupertinoTextField-class.html):**
  iOS tarzı metin giriş alanı.
- **[CupertinoSearchTextField](https://api.flutter.dev/flutter/cupertino/CupertinoSearchTextField-class.html):**
  iOS tarzı arama alanı.
- **[CupertinoCheckBox](https://www.google.com/search?q=https://api.flutter.dev/flutter/cupertino/CupertinoCheckBox-class.html):**
  macOS tarzı onay kutusu.
- **[CupertinoRadio](https://api.flutter.dev/flutter/cupertino/CupertinoRadio-class.html):**
  macOS tarzı radyo düğmesi.
- **[CupertinoSlider](https://api.flutter.dev/flutter/cupertino/CupertinoSlider-class.html):**
  Bir değer aralığından seçim yapmak için kullanılan kaydırıcı.
- **[CupertinoSwitch](https://api.flutter.dev/flutter/cupertino/CupertinoSwitch-class.html):**
  Tek bir ayarın açık/kapalı durumunu değiştirmek için kullanılan iOS
  tarzı anahtar.
- **[CupertinoDatePicker](https://api.flutter.dev/flutter/cupertino/CupertinoDatePicker-class.html):**
  iOS tarzı tarih veya tarih ve saat seçici.
- **[CupertinoTimerPicker](https://api.flutter.dev/flutter/cupertino/CupertinoTimerPicker-class.html):**
  iOS tarzı geri sayım zamanlayıcı seçicisi.
- **[CupertinoPicker](https://api.flutter.dev/flutter/cupertino/CupertinoPicker-class.html):**
  Kısa bir listeden bir öğe seçmek için kullanılan iOS tarzı tekerlekli
  seçici kontrolü.
- **[CupertinoSlidingSegmentedControl](https://api.flutter.dev/flutter/cupertino/CupertinoSlidingSegmentedControl-class.html):**
  Yatay bir listede birbirini dışlayan seçenekleri seçmek için
  kullanılan iOS 13 tarzı bölümlü kontrol.

## İletişim kutuları, menüler ve uyarılar (Dialogs, menus, and alerts)

Kullanıcıya bilgi vermek veya eylem seçenekleri sunmak için kullanılan
bileşenler.

- **[CupertinoAlertDialog](https://api.flutter.dev/flutter/cupertino/CupertinoAlertDialog-class.html):**
  iOS tarzı uyarı iletişim kutusu.
- **[CupertinoDialogAction](https://api.flutter.dev/flutter/cupertino/CupertinoDialogAction-class.html):**
  Genellikle `CupertinoAlertDialog` içinde kullanılan bir düğme.
- **[CupertinoActionSheet](https://api.flutter.dev/flutter/cupertino/CupertinoActionSheet-class.html):**
  Kullanıcının birçok seçenek arasından seçim yapmasını sağlayan, alttan
  kayarak gelen iOS tarzı eylem sayfası (modal).
- **[CupertinoActionSheetAction](https://api.flutter.dev/flutter/cupertino/CupertinoActionSheetAction-class.html):**
  Genellikle `CupertinoActionSheet` içinde kullanılan bir düğme.
- **[CupertinoContextMenu](https://api.flutter.dev/flutter/cupertino/CupertinoContextMenu-class.html):**
  Çocuğa uzun basıldığında açılan iOS tarzı tam ekran modlu rota.
  İçerikle ilgili eylemleri görüntülemek için kullanılır.
- **[CupertinoPopupSurface](https://api.flutter.dev/flutter/cupertino/CupertinoPopupSurface-class.html):**
  Uyarı iletişim kutusu veya eylem sayfası gibi iOS açılır yüzeylerine
  benzeyen yuvarlatılmış dikdörtgen yüzey.

## Formlar (Forms)

Form düzenleri oluşturmak için kullanılan widget’lar.

- **[CupertinoFormSection](https://api.flutter.dev/flutter/cupertino/CupertinoFormSection-class.html):**
  iOS tarzı form bölümü.
- **[CupertinoFormRow](https://api.flutter.dev/flutter/cupertino/CupertinoFormRow-class.html):**
  iOS tarzı form satırı.
- **[CupertinoTextFormFieldRow](https://api.flutter.dev/flutter/cupertino/CupertinoTextFormFieldRow-class.html):**
  Bir `CupertinoTextField`’ı saran `FormField` içeren bir
  `CupertinoFormRow` oluşturur.

## Listeler (Lists)

Liste görünümleri oluşturmak için kullanılan widget’lar.

- **[CupertinoListSection](https://api.flutter.dev/flutter/cupertino/CupertinoListSection-class.html):**
  Kaydırılabilir bir görünümü görüntülemek için iOS stilini kullanan
  kapsayıcı.
- **[CupertinoListTile](https://api.flutter.dev/flutter/cupertino/CupertinoListTile-class.html):**
  Listede bir satır oluşturmak için iOS stilini kullanan bir blok.

## Diğer (Miscellaneous)

- **[CupertinoActivityIndicator](https://api.flutter.dev/flutter/cupertino/CupertinoActivityIndicator-class.html):**
  iOS tarzı etkinlik göstergesi (dönen dairesel yükleyici).
- **[CupertinoColors](https://api.flutter.dev/flutter/cupertino/CupertinoColors-class.html):**
  iOS platform estetiğine uygun renkleri tanımlayan renk sabitleri
  paleti.
- **[CupertinoScrollbar](https://api.flutter.dev/flutter/cupertino/CupertinoScrollbar-class.html):**
  Kaydırılabilir bir widget’ın hangi kısmının görünür olduğunu gösteren
  iOS tarzı kaydırma çubuğu.
