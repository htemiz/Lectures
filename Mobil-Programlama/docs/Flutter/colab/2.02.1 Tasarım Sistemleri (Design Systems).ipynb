## 1. Tasarım Sistemlerine Giriş

### 🎯 Tasarım Sistemi Nedir?

**Tasarım sistemi**, bir uygulamanın veya ürün ailesinin tüm kullanıcı
arayüzü (UI) bileşenlerini, stil kurallarını ve etkileşim prensiplerini
standartlaştıran kapsamlı bir yapıdır.  
Amaç; uygulamanın her ekranında **tutarlılık**, **yeniden
kullanılabilirlik** ve **bakım kolaylığı** sağlamaktır.

Bir tasarım sistemi genellikle şunları içerir:

- 🎨 Renk paletleri  
- 🖋️ Tipografi kuralları  
- 📐 Boşluk ve hizalama sistemleri  
- 🧩 Hazır UI bileşenleri  
- 🔄 Etkileşim ve animasyon davranışları

------------------------------------------------------------------------

### 💡 Neden Tasarım Sistemi Kullanılır?

Tasarım sistemleri:

- Uygulamanın tüm ekranlarında görsel tutarlılık sağlar  
- Yeni özellik eklerken hız kazandırır  
- Tasarım ve yazılım ekipleri arasında ortak dil oluşturur  
- Büyük projelerde bakım maliyetini düşürür

------------------------------------------------------------------------

### 🧠 Flutter’da Tasarım Sistemleri

Flutter, yerleşik olarak iki ana tasarım sistemini destekler:

| Tasarım Sistemi | Platform Odaklı | Kullanılan Ana Widget |
|:----------------|:----------------|:----------------------|
| Material Design | Android / Web   | `MaterialApp`         |
| Cupertino       | iOS             | `CupertinoApp`        |

Bu iki sistem sayesinde Flutter ile **tek kod tabanından**, platforma
özgü görünümler oluşturmak mümkündür.

------------------------------------------------------------------------

### 🧪 Örnek: Basit Tasarım Sistemi Seçimi

#### Material Tabanlı Uygulama:

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'Material Tasarım Sistemi',
      home: Scaffold(
        appBar: AppBar(title: const Text('Material Örneği')),
        body: const Center(child: Text('Merhaba Material Tasarım Sistemi!')),
      ),
    );
  }
}
```

#### Cupertino Tabanlı Uygulama:

``` dart
import 'package:flutter/cupertino.dart';

void main() => runApp(const MyCupertinoApp());

class MyCupertinoApp extends StatelessWidget {
  const MyCupertinoApp({super.key});

  @override
  Widget build(BuildContext context) {
    return const CupertinoApp(
      home: CupertinoPageScaffold(
        navigationBar: CupertinoNavigationBar(
          middle: Text('Cupertino Örneği'),
        ),
        child: Center(child: Text('Merhaba Cupertino Tasarım Sistemi!')),
      ),
    );
  }
}
```

## 2. Material Design Sistemi

### 🎨 Material Design Nedir?

**Material Design**, Google tarafından geliştirilen modern bir tasarım
dilidir.  
Amaç; doğal hareketler, sezgisel etkileşimler ve tutarlı görsel yapı ile
kullanıcı deneyimini iyileştirmektir.

Flutter’da Material Design, varsayılan olarak gelen en kapsamlı tasarım
sistemidir.

------------------------------------------------------------------------

### 📐 Material Design Prensipleri

Material Design üç temel prensip üzerine kuruludur:

1.  **Material as a metaphor** – Arayüz öğeleri fiziksel yüzeyler gibi
    davranır.  
2.  **Bold, graphic, intentional** – Cesur renkler ve net hiyerarşi
    kullanılır.  
3.  **Motion provides meaning** – Animasyonlar anlamlı geçişler sunar.

------------------------------------------------------------------------

### 🧰 Flutter’da Material Bileşenleri

Flutter, yüzlerce hazır Material bileşeni sunar:

- `AppBar`
- `Scaffold`
- `ElevatedButton`
- `TextField`
- `Card`
- `NavigationBar`
- `FloatingActionButton`

Bu bileşenler otomatik olarak tema ile uyumlu çalışır.

------------------------------------------------------------------------

### 🌈 Material Theme (ThemeData)

Material tasarımının kalbi **ThemeData** sınıfıdır.  
Tüm uygulama genelinde renkleri, yazı tiplerini ve bileşen stillerini
buradan yönetiriz.

#### 🧪 Örnek: Özel Tema Tanımlama

``` dart
MaterialApp(
  theme: ThemeData(
    useMaterial3: true, // Material 3 aktif
    colorScheme: ColorScheme.fromSeed(
      seedColor: Colors.deepPurple,
    ),
    textTheme: const TextTheme(
      headlineMedium: TextStyle(fontSize: 28, fontWeight: FontWeight.bold),
      bodyMedium: TextStyle(fontSize: 16),
    ),
  ),
  home: const MyHomePage(),
);
```

------------------------------------------------------------------------

### 🆕 Material 3 (Material You)

Flutter artık **Material 3** tasarımını destekler.

Öne çıkan özellikler: - Dinamik renk şemaları  
- Daha yumuşak köşeler  
- Gelişmiş tipografi  
- Modern bileşenler

Material 3 kullanmak için:

``` dart
ThemeData(
  useMaterial3: true,
)
```

yazmak yeterlidir.

------------------------------------------------------------------------

### 🎯 Renk Şeması (ColorScheme)

Renkler `ColorScheme` ile merkezi olarak tanımlanır:

``` dart
colorScheme: ColorScheme.fromSeed(
  seedColor: Colors.teal,
)
```

Bu sayede tüm butonlar, arkaplanlar ve metinler uyumlu renklerde
görünür.

------------------------------------------------------------------------

### 🖋️ Tipografi (TextTheme)

Uygulamadaki tüm metin stilleri `TextTheme` ile belirlenir:

``` dart
textTheme: const TextTheme(
  titleLarge: TextStyle(fontSize: 22, fontWeight: FontWeight.w600),
  bodyMedium: TextStyle(fontSize: 16),
)
```

------------------------------------------------------------------------

### 🧪 Örnek: Tam Temalı Material Uygulama

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const MyApp());

class MyApp extends StatelessWidget {
  const MyApp({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'Material Tema Örneği',
      theme: ThemeData(
        useMaterial3: true,
        colorScheme: ColorScheme.fromSeed(seedColor: Colors.indigo),
      ),
      home: const MyHomePage(),
    );
  }
}

class MyHomePage extends StatelessWidget {
  const MyHomePage({super.key});

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text('Material 3 Demo')),
      floatingActionButton: FloatingActionButton(
        onPressed: () {},
        child: const Icon(Icons.add),
      ),
      body: const Center(
        child: Text('Merhaba Material Design!'),
      ),
    );
  }
}
```

------------------------------------------------------------------------

## 3. Cupertino (iOS Tasarım Sistemi)

### 🍏 Cupertino Tasarım Sistemi Nedir?

**Cupertino**, Flutter’ın **iOS işletim sistemine özgü** tasarım dilini
uygulamak için sunduğu widget koleksiyonudur.  
Apple’ın Human Interface Guidelines (HIG) kurallarına uygun arayüzler
oluşturmayı sağlar.

Eğer hedef kitlen iOS kullanıcılarıysa, Cupertino bileşenleri kullanarak
doğal iOS hissi veren uygulamalar geliştirebilirsin.

------------------------------------------------------------------------

### 📱 Cupertino ve Material Arasındaki Fark

| Özellik               | Material             | Cupertino              |
|:----------------------|:---------------------|:-----------------------|
| Kaynak                | Google               | Apple                  |
| Görünüm               | Android / Web odaklı | iOS odaklı             |
| Ana Uygulama Widget’ı | MaterialApp          | CupertinoApp           |
| Navigasyon            | AppBar               | CupertinoNavigationBar |
| Buton                 | ElevatedButton       | CupertinoButton        |

------------------------------------------------------------------------

### 🧰 Flutter’da Cupertino Temel Widget’ları

- `CupertinoApp`
- `CupertinoPageScaffold`
- `CupertinoNavigationBar`
- `CupertinoButton`
- `CupertinoTextField`
- `CupertinoSwitch`
- `CupertinoTabScaffold`

Bu bileşenler iOS görünümünü otomatik olarak uygular.

------------------------------------------------------------------------

### 🧪 Örnek: Basit Cupertino Uygulaması

``` dart
import 'package:flutter/cupertino.dart';

void main() => runApp(const MyCupertinoApp());

class MyCupertinoApp extends StatelessWidget {
  const MyCupertinoApp({super.key});

  @override
  Widget build(BuildContext context) {
    return const CupertinoApp(
      home: CupertinoPageScaffold(
        navigationBar: CupertinoNavigationBar(
          middle: Text('Cupertino Demo'),
        ),
        child: Center(
          child: Text('Merhaba iOS Tasarımı!'),
        ),
      ),
    );
  }
}
```

------------------------------------------------------------------------

### 🔘 Örnek: CupertinoButton

``` dart
CupertinoButton(
  child: const Text("iOS Buton"),
  onPressed: () {
    debugPrint("Cupertino butona basıldı!");
  },
)
```

Bu buton iOS stilinde görünür ve dokunma geri bildirimi sağlar.

------------------------------------------------------------------------

### 🔀 Material ve Cupertino’yu Birlikte Kullanma

Flutter, tek uygulama içinde iki tasarım sistemini birlikte kullanmaya
izin verir.

Örneğin:

``` dart
import 'package:flutter/material.dart';
import 'package:flutter/cupertino.dart';

class HybridPage extends StatelessWidget {
  const HybridPage({super.key});

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text("Material AppBar")),
      body: Center(
        child: CupertinoButton(
          child: const Text("Cupertino Buton"),
          onPressed: () {},
        ),
      ),
    );
  }
}
```

Bu sayede platforma göre widget seçimi yapabilirsin.

------------------------------------------------------------------------

### 📌 Platforma Göre Otomatik Widget Seçimi

``` dart
import 'dart:io';

Widget platformButton() {
  if (Platform.isIOS) {
    return CupertinoButton(
      child: const Text("iOS Button"),
      onPressed: () {},
    );
  } else {
    return ElevatedButton(
      child: const Text("Android Button"),
      onPressed: () {},
    );
  }
}
```

------------------------------------------------------------------------

## 4. Tema ve Stil Yönetimi

### 🎯 Tema Yönetimi Nedir?

Flutter’da **tema yönetimi**, uygulamanın tüm görsel stilini merkezi bir
yerden kontrol etmeyi sağlar.  
Renkler, yazı tipleri, buton stilleri ve arka planlar tek bir tema
tanımıyla yönetilir.

Bu yaklaşım: - Tutarlılık sağlar  
- Kod tekrarını azaltır  
- Tema değişimini kolaylaştırır

------------------------------------------------------------------------

### 🌗 Light ve Dark Tema

Flutter, açık (light) ve koyu (dark) temaları destekler.

``` dart
MaterialApp(
  theme: ThemeData.light(),
  darkTheme: ThemeData.dark(),
  themeMode: ThemeMode.system, // Sistem ayarına göre seç
);
```

Kullanıcının cihaz ayarına göre tema otomatik değişir.

------------------------------------------------------------------------

### 🌈 Özel Renk Paleti Oluşturma

``` dart
ThemeData(
  colorScheme: ColorScheme.fromSeed(
    seedColor: Colors.orange,
    brightness: Brightness.light,
  ),
);
```

Bu yöntem tüm uygulamaya uyumlu bir renk paleti üretir.

------------------------------------------------------------------------

### 🖋️ Global Yazı Tipi (Font) Yönetimi

``` dart
ThemeData(
  fontFamily: 'Roboto',
  textTheme: const TextTheme(
    titleLarge: TextStyle(fontSize: 24, fontWeight: FontWeight.bold),
    bodyMedium: TextStyle(fontSize: 16),
  ),
);
```

Tüm metinler otomatik olarak bu font ve stilleri kullanır.

------------------------------------------------------------------------

### 🎨 Bileşen Bazlı Stil Özelleştirme

Belirli widget’lar için tema seviyesinde özel stiller tanımlanabilir.

``` dart
ThemeData(
  elevatedButtonTheme: ElevatedButtonThemeData(
    style: ElevatedButton.styleFrom(
      backgroundColor: Colors.green,
      foregroundColor: Colors.white,
      shape: RoundedRectangleBorder(
        borderRadius: BorderRadius.circular(12),
      ),
    ),
  ),
);
```

Artık tüm `ElevatedButton` bileşenleri bu stile sahip olur.

------------------------------------------------------------------------

### 🔄 Dinamik Tema Değiştirme

Tema çalışma zamanında değiştirilebilir.

Basit örnek:

``` dart
class ThemeController extends ValueNotifier<ThemeMode> {
  ThemeController() : super(ThemeMode.light);

  void toggleTheme() {
    value = value == ThemeMode.light ? ThemeMode.dark : ThemeMode.light;
  }
}
```

Bu yapı, kullanıcıya “tema değiştir” seçeneği sunmak için kullanılır.

------------------------------------------------------------------------

### 🧪 Örnek: Tema Değiştirilebilir Uygulama

``` dart
void main() => runApp(const MyApp());

class MyApp extends StatefulWidget {
  const MyApp({super.key});

  @override
  State<MyApp> createState() => _MyAppState();
}

class _MyAppState extends State<MyApp> {
  ThemeMode _mode = ThemeMode.light;

  void _toggleTheme() {
    setState(() {
      _mode = _mode == ThemeMode.light ? ThemeMode.dark : ThemeMode.light;
    });
  }

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      theme: ThemeData.light(),
      darkTheme: ThemeData.dark(),
      themeMode: _mode,
      home: Scaffold(
        appBar: AppBar(title: const Text("Tema Yönetimi")),
        body: Center(
          child: ElevatedButton(
            onPressed: _toggleTheme,
            child: const Text("Tema Değiştir"),
          ),
        ),
      ),
    );
  }
}
```

------------------------------------------------------------------------

## 5. Özelleştirilmiş Bileşenler (Custom Components)

### 🧩 Özelleştirilmiş Bileşen Nedir?

Flutter’da **özelleştirilmiş bileşenler (custom components)**, tekrar
kullanılabilir UI parçaları oluşturmak için yazılan özel
widget’lardır.  
Tasarım sistemlerinde bu yaklaşım, tüm uygulamada **standart görünümlü
bileşenler** üretmek için kullanılır.

Örneğin:  
- Özel buton  
- Özel kart  
- Özel form alanı  
- Özel başlık widget’ı

------------------------------------------------------------------------

### 🎯 Neden Custom Component Kullanılır?

- Kod tekrarını azaltır  
- Tutarlı görünüm sağlar  
- Bakım ve güncelleme kolaylığı sunar  
- Tasarım sisteminin temel yapı taşlarını oluşturur

------------------------------------------------------------------------

### 🏗️ Basit Custom Button Örneği

Aşağıda, tasarım sistemine ait özel bir buton oluşturuyoruz.

``` dart
import 'package:flutter/material.dart';

class PrimaryButton extends StatelessWidget {
  final String text;
  final VoidCallback onPressed;

  const PrimaryButton({
    super.key,
    required this.text,
    required this.onPressed,
  });

  @override
  Widget build(BuildContext context) {
    return ElevatedButton(
      style: ElevatedButton.styleFrom(
        backgroundColor: Theme.of(context).colorScheme.primary,
        foregroundColor: Colors.white,
        padding: const EdgeInsets.symmetric(horizontal: 24, vertical: 12),
        shape: RoundedRectangleBorder(
          borderRadius: BorderRadius.circular(12),
        ),
      ),
      onPressed: onPressed,
      child: Text(text),
    );
  }
}
```

------------------------------------------------------------------------

### 🧪 Kullanımı

``` dart
PrimaryButton(
  text: "Kaydet",
  onPressed: () {
    debugPrint("Kaydet butonuna basıldı!");
  },
)
```

Artık uygulamanın her yerinde aynı görünüme sahip buton kullanılır.

------------------------------------------------------------------------

### 📦 Component Kütüphanesi Mantığı

Büyük projelerde custom component’ler genellikle şu klasör yapısında
tutulur:

    lib/
     └── design_system/
          ├── buttons/
          │    └── primary_button.dart
          ├── cards/
          │    └── info_card.dart
          ├── typography/
          │    └── app_text.dart
          └── colors/
               └── app_colors.dart

Bu yapı, gerçek bir tasarım sistemi kütüphanesi oluşturur.

------------------------------------------------------------------------

### 🃏 Örnek: Özel Kart Bileşeni

``` dart
class InfoCard extends StatelessWidget {
  final String title;
  final String description;

  const InfoCard({
    super.key,
    required this.title,
    required this.description,
  });

  @override
  Widget build(BuildContext context) {
    return Card(
      elevation: 4,
      child: Padding(
        padding: const EdgeInsets.all(16),
        child: Column(
          crossAxisAlignment: CrossAxisAlignment.start,
          children: [
            Text(title, style: Theme.of(context).textTheme.titleLarge),
            const SizedBox(height: 8),
            Text(description),
          ],
        ),
      ),
    );
  }
}
```

## 6. Responsive Tasarım

### 📱 Responsive Tasarım Nedir?

**Responsive tasarım**, uygulamanın farklı ekran boyutlarına (telefon,
tablet, masaüstü, web) uyumlu şekilde görünmesini sağlar.  
Flutter, tek kod tabanıyla çoklu ekran boyutlarına uyum sağlamayı
kolaylaştırır.

------------------------------------------------------------------------

### 🎯 Neden Responsive Tasarım?

- Mobil, tablet ve web için tek uygulama  
- Daha iyi kullanıcı deneyimi  
- Ekran alanının verimli kullanımı  
- Geleceğe dönük tasarım yaklaşımı

------------------------------------------------------------------------

### 📏 MediaQuery ile Ekran Boyutu Alma

``` dart
double width = MediaQuery.of(context).size.width;
double height = MediaQuery.of(context).size.height;
```

Bu bilgilerle ekrana göre farklı UI’lar üretebilirsin.

------------------------------------------------------------------------

### 🧱 LayoutBuilder Kullanımı

`LayoutBuilder`, mevcut alanın genişliğine göre widget üretmeyi sağlar.

``` dart
LayoutBuilder(
  builder: (context, constraints) {
    if (constraints.maxWidth < 600) {
      return const Text("Mobil Görünüm");
    } else {
      return const Text("Tablet / Web Görünüm");
    }
  },
);
```

------------------------------------------------------------------------

### 🧩 Breakpoint Mantığı

Genelde şu kırılım noktaları (breakpoints) kullanılır:

| Cihaz        | Genişlik      |
|:-------------|:--------------|
| Telefon      | \< 600 px     |
| Tablet       | 600 – 1024 px |
| Masaüstü/Web | \> 1024 px    |

------------------------------------------------------------------------

### 🧪 Örnek: Responsive Sayfa Tasarımı

``` dart
class ResponsivePage extends StatelessWidget {
  const ResponsivePage({super.key});

  @override
  Widget build(BuildContext context) {
    return Scaffold(
      appBar: AppBar(title: const Text("Responsive Demo")),
      body: LayoutBuilder(
        builder: (context, constraints) {
          if (constraints.maxWidth < 600) {
            return const Center(child: Text("📱 Mobil Görünüm"));
          } else if (constraints.maxWidth < 1024) {
            return const Center(child: Text("📱 Tablet Görünüm"));
          } else {
            return const Center(child: Text("🖥️ Masaüstü / Web Görünüm"));
          }
        },
      ),
    );
  }
}
```

------------------------------------------------------------------------

### 🌐 Grid ve Flexible Layout

Responsive tasarımda `Row`, `Column`, `Expanded` ve `GridView` çok
kullanılır.

``` dart
GridView.count(
  crossAxisCount: 2,
  children: const [
    Card(child: Center(child: Text("Kutu 1"))),
    Card(child: Center(child: Text("Kutu 2"))),
  ],
);
```

## 7. Erişilebilirlik (Accessibility)

### ♿ Erişilebilirlik Nedir?

**Erişilebilirlik (Accessibility)**, uygulamanın görme, işitme, motor
veya bilişsel farklılıklara sahip kullanıcılar tarafından da rahatça
kullanılabilmesini sağlar.

İyi bir tasarım sistemi: - Herkes için kullanılabilir olmalı  
- Ekran okuyucularla uyumlu çalışmalı  
- Yeterli renk kontrastı sunmalı  
- Metin ölçeklendirmeyi desteklemeli

Flutter, erişilebilir uygulamalar geliştirmek için güçlü araçlar sunar.

------------------------------------------------------------------------

### 🎯 Neden Önemlidir?

- Daha geniş kullanıcı kitlesine ulaşılır  
- Yasal erişilebilirlik standartlarına uyum sağlanır  
- Daha kaliteli kullanıcı deneyimi sunulur

------------------------------------------------------------------------

### 🔊 Semantics Widget

Flutter’da ekran okuyucuların (TalkBack, VoiceOver) içeriği anlaması
için **Semantics** widget’ı kullanılır.

``` dart
Semantics(
  label: "Kaydet butonu",
  hint: "Formu kaydetmek için çift dokunun",
  child: ElevatedButton(
    onPressed: () {},
    child: const Text("Kaydet"),
  ),
);
```

Bu sayede ekran okuyucu butonu doğru şekilde tanımlar.

------------------------------------------------------------------------

### 🔍 Görsel Kontrast

Metin ve arka plan arasında yeterli kontrast olmalıdır.

``` dart
Text(
  "Okunabilir Metin",
  style: TextStyle(
    color: Colors.black,
    backgroundColor: Colors.white,
  ),
);
```

Tasarım sisteminde renk paleti seçerken kontrast oranına dikkat
edilmelidir.

------------------------------------------------------------------------

### 🔠 Metin Ölçeklendirme Desteği

Kullanıcı cihaz ayarlarından yazı boyutunu büyüttüğünde uygulama uyumlu
olmalıdır.

``` dart
Text(
  "Ölçeklenebilir Metin",
  textScaleFactor: MediaQuery.of(context).textScaleFactor,
);
```

Flutter varsayılan olarak metin ölçeklendirmeyi destekler.

------------------------------------------------------------------------

### 🧭 Odak Yönetimi (Focus)

Klavye veya erişilebilirlik cihazları ile gezinti için **Focus** yapısı
kullanılır.

``` dart
FocusTraversalGroup(
  child: Column(
    children: [
      TextField(),
      ElevatedButton(onPressed: () {}, child: const Text("Gönder")),
    ],
  ),
);
```

Bu sayede odak sırası doğru şekilde yönetilir.

------------------------------------------------------------------------

### 📢 Erişilebilirlik Testi

Flutter’da erişilebilirlik denetimi için:

- Android: TalkBack  
- iOS: VoiceOver  
- Flutter DevTools: Accessibility Inspector

kullanılabilir.
