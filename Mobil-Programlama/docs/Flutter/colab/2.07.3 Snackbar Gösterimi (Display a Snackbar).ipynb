# Snackbar Gösterimi

Kullanıcılarınıza belirli eylemler gerçekleştiğinde (örneğin bir mesajın
silinmesi, bir işlemin tamamlanması vb.) kısa süreli bilgilendirmeler
yapmak isteyebilirsiniz. Material Design’da bu işlev **`SnackBar`**
widget’ı ile gerçekleştirilir.

Bu işlemi 3 adımda uygulayacağız:

1.  Bir `Scaffold` oluşturmak.
2.  `SnackBar`’ı görüntülemek.
3.  İsteğe bağlı bir eylem (aksiyon) eklemek.

------------------------------------------------------------------------

## 1. `Scaffold` Oluşturma

Material Design yönergelerine uyan uygulamalar, tutarlı bir görsel
yapıya sahip olmalıdır. `SnackBar`, genellikle ekranın alt kısmında
belirir ve `FloatingActionButton` gibi diğer önemli widget’larla
çakışmamalıdır.

Bunu sağlamak için **`Scaffold`** widget’ı kullanılır. `Scaffold`, bu
görsel yapıyı kurar ve widget’ların üst üste binmesini engeller.

``` dart
return MaterialApp(
  title: 'SnackBar Demo',
  home: Scaffold(
    appBar: AppBar(title: const Text('SnackBar Demo')),
    body: const SnackBarPage(), // Sayfa içeriği
  ),
);
```

------------------------------------------------------------------------

## 2. `SnackBar` Görüntüleme

`Scaffold` hazır olduktan sonra, bir `SnackBar` oluşturup
gösterebilirsiniz. Bu işlem için **`ScaffoldMessenger`** kullanılır.

1.  Önce `SnackBar` nesnesini tanımlayın.
2.  Sonra `ScaffoldMessenger.of(context).showSnackBar()` metodunu
    çağırarak ekrana getirin.

``` dart
// Gösterilecek SnackBar'ı tanımla
const snackBar = SnackBar(content: Text('Yaşasın! Bir SnackBar!'));

// Widget ağacında ScaffoldMessenger'ı bul ve SnackBar'ı göster
ScaffoldMessenger.of(context).showSnackBar(snackBar);
```

------------------------------------------------------------------------

## 3. İsteğe Bağlı Eylem (Action) Ekleme

Kullanıcıya sadece bilgi vermekle kalmayıp, bir eylem (aksiyon) sunmak
da isteyebilirsiniz. Örneğin, yanlışlıkla silinen bir mesajı geri almak
için bir “Geri Al” (Undo) butonu koyabilirsiniz.

Bunun için `SnackBar` widget’ının **`action`** özelliğine bir
**`SnackBarAction`** widget’ı eklemeniz yeterlidir.

``` dart
final snackBar = SnackBar(
  content: const Text('Mesaj silindi'),
  action: SnackBarAction(
    label: 'Geri Al', // Buton metni
    onPressed: () {
      // Geri alma işlemi için kodlar buraya
      print('İşlem geri alındı!');
    },
  ),
);

ScaffoldMessenger.of(context).showSnackBar(snackBar);
```

### Özet

- **Amaç:** Kullanıcıyı kısa süreliğine bilgilendirmek.
- **Konum:** Ekranın alt kısmı (`Scaffold` tarafından yönetilir).
- **Yönetici:** `ScaffoldMessenger`.
- **Etkileşim:** `SnackBarAction` ile buton eklenebilir.

## İnteraktif Örnek

``` dart
import 'package:flutter/material.dart';

void main() => runApp(const SnackBarDemo());

class SnackBarDemo extends StatelessWidget {
  const SnackBarDemo({super.key});

  @override
  Widget build(BuildContext context) {
    return MaterialApp(
      title: 'SnackBar Demo',
      home: Scaffold(
        appBar: AppBar(title: const Text('SnackBar Demo')),
        body: const SnackBarPage(),
      ),
    );
  }
}

class SnackBarPage extends StatelessWidget {
  const SnackBarPage({super.key});

  @override
  Widget build(BuildContext context) {
    return Center(
      child: ElevatedButton(
        onPressed: () {
          final snackBar = SnackBar(
            content: const Text('Yay! A SnackBar!'),
            action: SnackBarAction(
              label: 'Undo',
              onPressed: () {
                // Some code to undo the change.
              },
            ),
          );

          // Find the ScaffoldMessenger in the widget tree
          // and use it to show a SnackBar.
          ScaffoldMessenger.of(context).showSnackBar(snackBar);
        },
        child: const Text('Show SnackBar'),
      ),
    );
  }
}
```
