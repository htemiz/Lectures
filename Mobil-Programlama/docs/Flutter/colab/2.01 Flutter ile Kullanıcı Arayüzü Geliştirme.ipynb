# Flutter ile Kullanıcı Arayüzü Geliştirme

## Giriş

Flutter widget’ları, React’ten ilham alan modern bir framework üzerine
inşa edilmiştir. Temel fikir, kullanıcı arayüzünüzü widget’lardan
oluşturmanızdır. Widget’lar, mevcut yapılandırmalarına ve durumlarına
(state) göre görünümlerinin nasıl olması gerektiğini tanımlar. Bir
widget’ın durumu değiştiğinde, widget kendi tanımını yeniden oluşturur.
Framework, bu yeni tanımı önceki tanımla karşılaştırır ve bir durumdan
diğerine geçiş için render ağacında yapılması gereken en küçük
değişiklikleri uygular.

Flutter’da kullanıcı arayüzü (UI) geliştirme, **deklaratif
(bildirimsel)** bir yaklaşıma dayanır.  
Bu yaklaşımda, “nasıl çizileceğini” değil, “nasıl görünmesi gerektiğini”
tarif edersiniz.

Flutter widget’ları, React’ten ilham alan modern bir framework üzerine
inşa edilmiştir.  
Temel fikir şudur:

> Kullanıcı arayüzünüz tamamen widget’lardan oluşur.

Bir widget, mevcut **yapılandırmasına (configuration)** ve **durumuna
(state)** göre ekranda nasıl görünmesi gerektiğini tanımlar.

Bir widget’ın durumu değiştiğinde: 1. Widget yeniden `build()` edilir.
2. Framework yeni widget ağacını öncekiyle karşılaştırır (diff işlemi).
3. Render ağacında sadece gerekli minimum değişiklikler yapılır.

Bu mekanizma Flutter’a **yüksek performans** ve **akıcı animasyonlar**
kazandırır.

> **Not:**  
> Flutter’a kod yazarak daha yakından aşina olmak isterseniz, düzen
> (layout) oluşturma ve uygulamanıza etkileşim ekleme konularına göz
> atabilirsiniz.

## Hello World

En küçük Flutter uygulaması yalnızca `runApp()` fonksiyonunu çağırır:

``` dart
import 'package:flutter/material.dart';

void main() {
  runApp(const Center(
    child: Text(
      'Hello, world',
      textDirection: TextDirection.ltr,
    ),
  ));
}
```

`runApp()` fonksiyonu verilen widget’ı widget ağacının kökü yapar.
`MaterialApp` kullanıldığında `textDirection` otomatik yönetilir.

`runApp()` fonksiyonu verilen Widget’ı widget ağacının kökü yapar. Bu
örnekte widget ağacı iki widget’tan oluşur: `Center` widget’ı ve onun
altındaki `Text` widget’ı. Framework kök widget’ın ekranı tamamen
kaplamasını sağlar, böylece “Hello, world” metni ekranın ortasında
görünür. Bu örnekte metin yönü (textDirection) ayrıca belirtilmelidir.
`MaterialApp` widget’ı kullanıldığında bu işlem otomatik olarak yapılır.

Uygulama geliştirirken genellikle `StatelessWidget` veya
`StatefulWidget` sınıflarından türeyen yeni widget’lar yazarsınız. Bir
widget’ın temel görevi `build()` fonksiyonunu uygulamaktır. Bu
fonksiyon, widget’ı daha alt seviye widget’lar cinsinden tanımlar.
Framework bu süreci, en altta RenderObject’leri temsil eden widget’lara
ulaşana kadar devam ettirir.

------------------------------------------------------------------------

## StatelessWidget ve StatefulWidget

Flutter’da yeni widget’lar genellikle iki temel sınıftan türetilir:

### StatelessWidget

- Durum (state) tutmaz.
- İçeriği sadece aldığı parametrelere göre çizilir.
- Sabit içerikler için idealdir.

### StatefulWidget

- İçinde değişebilen durum (state) barındırır.
- Kullanıcı etkileşimi veya veri değişimi sonrası yeniden çizilir.

Her widget’ın ana görevi `build()` fonksiyonunu tanımlamaktır:

``` dart
@override
Widget build(BuildContext context) {
  return Text("Merhaba");
}
```

Framework, `build()` metodunu çağırarak widget’ın görünümünü oluşturur.

## Temel Widget’lar

Flutter, güçlü temel widget’lar sunar. En yaygın olanlar:

### Text

`Text` widget’ı, uygulamanız içinde stillendirilmiş metin oluşturmanızı
sağlar.

``` dart
Text(
  'Merhaba Flutter',
  style: TextStyle(fontSize: 24),
);
```

### Row ve Column

Bu esnek widget’lar yatay (`Row`) ve dikey (`Column`) yönlerde düzen
oluşturmanızı sağlar. Tasarımları web’deki flexbox modeline dayanır.

``` dart
Row(
  children: [
    Icon(Icons.star),
    Text("Yıldız"),
  ],
);
```

``` dart
Column(
  children: [
    Text("Satır 1"),
    Text("Satır 2"),
  ],
);
```

> Bu yapı web’deki **flexbox** düzenine benzer.

### Stack

`Stack`, widget’ları üst üste yerleştirmenizi sağlar. `Positioned`
widget’ı ile çocuk widget’ları üst, alt, sağ veya sol kenara göre
konumlandırabilirsiniz. Web’deki absolute positioning modeline benzer.

``` dart
Stack(
  children: [
    Container(width: 100, height: 100, color: Colors.blue),
    Positioned(
      top: 10,
      left: 10,
      child: Text("Üstte"),
    ),
  ],
);
```

### Container

`Container` dikdörtgen bir görsel öğe oluşturur. Arka plan, kenarlık
veya gölge gibi `BoxDecoration` özellikleri eklenebilir. Margin, padding
ve boyut kısıtlamaları uygulanabilir. Ayrıca 3B dönüşümler yapılabilir.

``` dart
Container(
  padding: EdgeInsets.all(16),
  decoration: BoxDecoration(
    color: Colors.orange,
    borderRadius: BorderRadius.circular(8),
  ),
  child: Text("Kutucuk"),
);
```

> Container; padding, margin, arka plan, gölge ve dönüşüm eklemek için
> kullanılır.

------------------------------------------------------------------------

## Material Design Kullanımı

Flutter, Google’ın Material Design sistemini destekler.

`pubspec.yaml` dosyanızda aşağıdaki satırın bulunması önerilir:

``` yaml
flutter:
  uses-material-design: true
```

Bu satır Material ikonlarını kullanabilmenizi sağlar.

Ana uygulama:

``` dart
void main() {
  runApp(MaterialApp(
    home: Scaffold(
      appBar: AppBar(title: Text("Başlık")),
      body: Center(child: Text("Merhaba")),
      floatingActionButton: FloatingActionButton(
        onPressed: () {},
        child: Icon(Icons.add),
      ),
    ),
  ));
}
```

Birçok Material Design widget’ı, tema verilerini alabilmek için
`MaterialApp` içinde çalıştırılmalıdır.

`Scaffold` widget’ı sayfanın temel düzenini oluşturur. `AppBar`,
`FloatingActionButton` gibi bileşenleri kolayca eklemenizi sağlar.

Widget’ların diğer widget’lara argüman olarak geçirilmesi, Flutter’daki
güçlü tasarım kalıplarından biridir.

> **Not:**  
> Flutter’da iki temel tasarım sistemi vardır: Material ve Cupertino.
> iOS görünümü için `Cupertino` bileşenleri kullanılabilir.

------------------------------------------------------------------------

## Dokunma (Gestures)

Kullanıcı etkileşimini algılamak için `GestureDetector` kullanılır. Bu
widget görsel değildir, ancak dokunma, sürükleme ve ölçekleme gibi
hareketleri algılar.

Örneğin `onTap()` ile tıklama olayları yakalanabilir.

`ElevatedButton`, `IconButton` ve `FloatingActionButton` gibi
widget’lar, dahili olarak `GestureDetector` kullanır.

``` dart
GestureDetector(
  onTap: () {
    print("Tıklandı!");
  },
  child: Container(
    padding: EdgeInsets.all(20),
    color: Colors.green,
    child: Text("Bana Dokun"),
  ),
);
```

- `onTap` → Kullanıcı dokunduğunda çalışır.
- Ayrıca `onDoubleTap`, `onLongPress`, `onPanUpdate` gibi birçok gesture
  vardır.

Buton widget’ları (ElevatedButton, IconButton vb.) zaten GestureDetector
içerir.

## Girdi Sonrası Widget Güncelleme - StatefulWidget ve setState()

Şimdiye kadar stateless widget’lar kullandık. Ancak etkileşimli
uygulamalarda durum (state) tutmak gerekir. Bu amaçla `StatefulWidget`
kullanılır.

`StatefulWidget` → Durum (state) bilgisini saklar.  
`State` → Kalıcıdır ve `build()` çağrıları arasında korunur.

Durum değiştiğinde `setState()` çağrılır. Bu, widget’ın yeniden
çizilmesini sağlar.

Eğer `setState()` çağrılmazsa, arayüz değişiklikleri ekrana yansıtılmaz.

``` dart
class CounterApp extends StatefulWidget {
  @override
  _CounterAppState createState() => _CounterAppState();
}

class _CounterAppState extends State<CounterApp> {
  int counter = 0;

  void increment() {
    setState(() {
      counter++;
    });
  }

  @override
  Widget build(BuildContext context) {
    return Column(
      mainAxisAlignment: MainAxisAlignment.center,
      children: [
        Text("Sayaç: $counter"),
        ElevatedButton(
          onPressed: increment,
          child: Text("Arttır"),
        ),
      ],
    );
  }
}
```

- `counter` değişkeni state’tir.
- `setState()` çağrılınca framework bu widget’ı yeniden çizer.

> `setState()` çağrılmazsa, ekranda değişiklik görünmez.

------------------------------------------------------------------------

## Alışveriş Listesi Örneği (State Yönetimi)

`ShoppingListItem` adlı bir stateless widget düşünelim. Bu widget, ürün
sepetteyse farklı, değilse farklı görünür.

Kullanıcı bir ürüne dokunduğunda widget doğrudan kendi durumunu
değiştirmez. Bunun yerine, üst widget’a bir callback gönderir. Böylece
durum, üst seviyede saklanır ve daha uzun süre korunur.

Üst widget (`ShoppingList`) bir `StatefulWidget`’tır ve sepet bilgisini
saklar. Durum değiştiğinde `setState()` çağrılır ve liste yeniden
çizilir.

Framework, eski ve yeni widget ağaçlarını karşılaştırır ve sadece
gerekli farkları uygular. Bu nedenle performans yüksektir.

``` dart
class ShoppingList extends StatefulWidget {
  @override
  _ShoppingListState createState() => _ShoppingListState();
}

class _ShoppingListState extends State<ShoppingList> {
  final List<String> _cart = [];

  void _toggle(String item) {
    setState(() {
      if (_cart.contains(item)) {
        _cart.remove(item);
      } else {
        _cart.add(item);
      }
    });
  }

  @override
  Widget build(BuildContext context) {
    final items = ["Elma", "Armut", "Muz"];

    return ListView(
      children: items.map((item) {
        final inCart = _cart.contains(item);
        return ListTile(
          title: Text(item),
          trailing: Icon(
            inCart ? Icons.check_box : Icons.check_box_outline_blank,
          ),
          onTap: () => _toggle(item),
        );
      }).toList(),
    );
  }
}
```

- Ürünler sepete eklenir/çıkarılır.
- State üst widget’ta tutulur.
- Çocuk widget’lar sadece görünümü temsil eder.

Bu yaklaşım Flutter’da **tek yönlü veri akışı** sağlar.

## Widget Yaşam Döngüsü (Lifecycle)

- `createState()` → State nesnesi oluşturulur.
- `initState()` → Bir kere çalışan başlatma fonksiyonu.
- `dispose()` → Widget kaldırıldığında temizlik yapılır.

Örneğin: - `initState()` → Animasyon başlatma - `dispose()` → Timer veya
servis aboneliklerini kapatma.

``` dart
@override
void initState() {
  super.initState();
  print("Başlatıldı");
}

@override
void dispose() {
  print("Temizlendi");
  super.dispose();
}
```

------------------------------------------------------------------------

## Keys

Varsayılan olarak framework widget’ları sıralarına göre eşleştirir.  
`Key` kullanarak hangi widget’ın hangisiyle eşleşeceğini kontrol
edebilirsiniz.

Özellikle listelerde performans ve doğru durum eşlemesi için önemlidir.

``` dart
ListView(
  children: items.map((item) {
    return ListTile(
      key: ValueKey(item),
      title: Text(item),
    );
  }).toList(),
);
```

``` dart
final GlobalKey<FormState> formKey = GlobalKey<FormState>();
```
